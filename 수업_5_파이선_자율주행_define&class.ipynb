```python

# ===== 1Îã®Í≥Ñ: Í∏∞Î≥∏ ÏûêÏú®Ï£ºÌñâÏ∞® ÌÅ¥ÎûòÏä§ =====
print("=== 1Îã®Í≥Ñ: Í∏∞Î≥∏ ÏûêÏú®Ï£ºÌñâÏ∞® ===")

class BasicCar:
    """Í∞ÄÏû• Í∏∞Î≥∏Ï†ÅÏù∏ ÏûêÏú®Ï£ºÌñâÏ∞®"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = 0
        print(f"üöó {self.name} ÏÉùÏÑ±!")

    def move(self, distance):
        if self.battery >= distance:
            self.battery -= distance
            self.position += distance
            print(f"‚úÖ {distance}km Ïù¥Îèô ÏôÑÎ£å! ÏúÑÏπò: {self.position}km, Î∞∞ÌÑ∞Î¶¨: {self.battery}%")
        else:
            print(f"‚ùå Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°±! ÌòÑÏû¨: {self.battery}%, ÌïÑÏöî: {distance}%")

    def charge(self, amount):
        self.battery += amount
        if self.battery > 100:
            self.battery = 100
        print(f"üîã Ï∂©Ï†Ñ ÏôÑÎ£å! Î∞∞ÌÑ∞Î¶¨: {self.battery}%")

# 1Îã®Í≥Ñ ÌÖåÏä§Ìä∏
car1 = BasicCar("ÌÖåÏä¨Îùº")
car1.move(30)
car1.move(50)
car1.charge(40)
car1.move(50)

print("\n" + "="*50 + "\n")
```

```python
# ===== 2Îã®Í≥Ñ: ÏÑºÏÑú Í∏∞Îä• Ï∂îÍ∞Ä =====
print("=== 2Îã®Í≥Ñ: ÏÑºÏÑú Í∏∞Îä• Ï∂îÍ∞Ä ===")

class SensorCar:
    """ÏÑºÏÑúÍ∞Ä ÏûàÎäî ÏûêÏú®Ï£ºÌñâÏ∞®"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = 0
        self.sensors = {
            'camera': True,
            'lidar': True,
            'radar': True
        }
        print(f"üöó {self.name} (ÏÑºÏÑú Ïû•Ï∞©) ÏÉùÏÑ±!")

    def check_sensors(self):
        print(f"üîç {self.name} ÏÑºÏÑú Ï†êÍ≤Ä:")
        working_count = 0

        for sensor_name, is_working in self.sensors.items():
            status = "Ï†ïÏÉÅ" if is_working else "Í≥†Ïû•"
            print(f"  {sensor_name}: {status}")
            if is_working:
                working_count += 1

        if working_count >= 2:
            print("‚úÖ Ï£ºÌñâ Í∞ÄÎä• (ÏÑºÏÑú 2Í∞ú Ïù¥ÏÉÅ Ï†ïÏÉÅ)")
            return True
        else:
            print("‚ùå Ï£ºÌñâ Î∂àÍ∞Ä (ÏÑºÏÑú Î∂ÄÏ°±)")
            return False

    def move_safely(self, distance):
        if self.check_sensors():
            if self.battery >= distance:
                self.battery -= distance
                self.position += distance
                print(f"‚úÖ ÏïàÏ†ÑÌïòÍ≤å {distance}km Ïù¥Îèô! ÏúÑÏπò: {self.position}km")
            else:
                print(f"‚ùå Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°±!")
        else:
            print("‚ö†Ô∏è ÏÑºÏÑú Ïù¥ÏÉÅÏúºÎ°ú Ïù¥Îèô Ï§ëÎã®!")

# 2Îã®Í≥Ñ ÌÖåÏä§Ìä∏
car2 = SensorCar("ÌòÑÎåÄ")
car2.move_safely(20)

# ÏÑºÏÑú Í≥†Ïû• ÏãúÎÆ¨Î†àÏù¥ÏÖò
car2.sensors['camera'] = False
car2.sensors['lidar'] = False
car2.move_safely(20)

print("\n" + "="*50 + "\n")
```

```python
# ===== 3Îã®Í≥Ñ: Ïã†Ìò∏Îì± ÏãúÏä§ÌÖú Ï∂îÍ∞Ä =====
print("=== 3Îã®Í≥Ñ: Ïã†Ìò∏Îì± ÏãúÏä§ÌÖú ===")

class TrafficLight:
    """Ïã†Ìò∏Îì± ÌÅ¥ÎûòÏä§"""

    def __init__(self, name, initial_color='red'):
        self.name = name
        self.color = initial_color
        self.time_left = 30
        print(f"üö¶ {self.name} Ïã†Ìò∏Îì± ÏÑ§Ïπò ({self.color})")

    def get_status(self):
        if self.color == 'green':
            return True  # ÌÜµÍ≥º Í∞ÄÎä•
        else:
            return False  # Ï†ïÏßÄ

    def change_signal(self):
        if self.color == 'red':
            self.color = 'green'
        elif self.color == 'green':
            self.color = 'yellow'
        else:
            self.color = 'red'
        print(f"üö¶ {self.name}: {self.color} Ïã†Ìò∏Î°ú Î≥ÄÍ≤Ω")

class SmartCar:
    """Ïã†Ìò∏Îì±ÏùÑ Ïù∏ÏãùÌïòÎäî ÏûêÏú®Ï£ºÌñâÏ∞®"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = 0
        print(f"üöó {self.name} (Ïä§ÎßàÌä∏) ÏÉùÏÑ±!")

    def approach_intersection(self, traffic_light):
        print(f"\nüöó {self.name}: {traffic_light.name} Ï†ëÍ∑º")
        print(f"üö¶ ÌòÑÏû¨ Ïã†Ìò∏: {traffic_light.color}")

        if traffic_light.get_status():
            print("‚úÖ Ï¥àÎ°ùÎ∂à - ÌÜµÍ≥º!")
            self.position += 10
            self.battery -= 5
        else:
            print("üõë Îπ®Í∞ÑÎ∂à/ÎÖ∏ÎûÄÎ∂à - Ï†ïÏßÄ!")
            print("‚è≥ Ïã†Ìò∏ Î≥ÄÍ≤Ω ÎåÄÍ∏∞ Ï§ë...")

# 3Îã®Í≥Ñ ÌÖåÏä§Ìä∏
light1 = TrafficLight("ÍµêÏ∞®Î°úA", 'red')
car3 = SmartCar("BMW")

car3.approach_intersection(light1)
light1.change_signal()  # Ï¥àÎ°ùÎ∂àÎ°ú Î≥ÄÍ≤Ω
car3.approach_intersection(light1)

print("\n" + "="*50 + "\n")
```

```python
# ===== 4Îã®Í≥Ñ: Í≤ΩÎ°ú Í≥ÑÌöç Í∏∞Îä• =====
print("=== 4Îã®Í≥Ñ: Í≤ΩÎ°ú Í≥ÑÌöç ===")

class NavigationCar:
    """Í≤ΩÎ°ú Í≥ÑÌöçÏù¥ Í∞ÄÎä•Ìïú ÏûêÏú®Ï£ºÌñâÏ∞®"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = [0, 0]  # [x, y] Ï¢åÌëú
        print(f"üöó {self.name} (ÎÇ¥ÎπÑÍ≤åÏù¥ÏÖò) ÏÉùÏÑ±!")

    def plan_route(self, waypoints):
        print(f"\nüó∫Ô∏è {self.name} Í≤ΩÎ°ú Í≥ÑÌöç:")
        total_distance = 0

        current_x, current_y = self.position

        for i, waypoint in enumerate(waypoints):
            target_x, target_y = waypoint
            distance = abs(target_x - current_x) + abs(target_y - current_y)
            total_distance += distance
            print(f"  {i+1}Îã®Í≥Ñ: ({target_x}, {target_y}) - Í±∞Î¶¨ {distance}km")
            current_x, current_y = target_x, target_y

        print(f"üìè Ï¥ù Í±∞Î¶¨: {total_distance}km")
        battery_needed = total_distance * 2

        if self.battery >= battery_needed:
            print("‚úÖ Î∞∞ÌÑ∞Î¶¨ Ï∂©Î∂Ñ - Ï∂úÎ∞ú Í∞ÄÎä•!")
            return True
        else:
            print(f"‚ùå Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°± - ÌïÑÏöî: {battery_needed}%, ÌòÑÏû¨: {self.battery}%")
            return False

    def move_to_waypoint(self, x, y):
        start_x, start_y = self.position
        distance = abs(x - start_x) + abs(y - start_y)
        battery_cost = distance * 2

        if self.battery >= battery_cost:
            self.position = [x, y]
            self.battery -= battery_cost
            print(f"‚úÖ ({x}, {y}) ÎèÑÏ∞©! Î∞∞ÌÑ∞Î¶¨: {self.battery}%")
            return True
        else:
            print(f"‚ùå Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°±ÏúºÎ°ú Ïù¥Îèô Ïã§Ìå®!")
            return False

# 4Îã®Í≥Ñ ÌÖåÏä§Ìä∏
car4 = NavigationCar("Íµ¨Í∏ÄÏπ¥")
route = [(3, 0), (3, 5), (8, 5), (8, 10)]

if car4.plan_route(route):
    print("\nüõ£Ô∏è Í≤ΩÎ°ú Ïã§Ìñâ:")
    for waypoint in route:
        x, y = waypoint
        if not car4.move_to_waypoint(x, y):
            break

print("\n" + "="*50 + "\n")
```

```python
# ===== 5Îã®Í≥Ñ: Í∞ÑÎã®Ìïú Ìï®ÏàòÎì§ =====
print("=== 5Îã®Í≥Ñ: Ïú†Ïö©Ìïú Ìï®ÏàòÎì§ ===")

def check_weather_condition(weather):
    """ÎÇ†Ïî®Ïóê Îî∞Î•∏ Ï£ºÌñâ Í∂åÏû•ÏÇ¨Ìï≠"""
    print(f"üå§Ô∏è ÌòÑÏû¨ ÎÇ†Ïî®: {weather}")

    if weather == "ÎßëÏùå":
        print("‚úÖ Ï†ïÏÉÅ Ïö¥Ìñâ Í∞ÄÎä•")
        return 100  # ÏµúÎåÄ ÏÜçÎèÑ
    elif weather == "ÎπÑ":
        print("‚ö†Ô∏è ÏÑúÌñâ Ïö¥Ï†Ñ Í∂åÏû•")
        return 60   # ÏÜçÎèÑ Ï†úÌïú
    elif weather == "Îàà":
        print("üî∂ Îß§Ïö∞ Ï£ºÏùò Ïö¥Ï†Ñ")
        return 40   # Ï†ÄÏÜç Ïö¥Ìñâ
    else:
        print("üö´ Ïö¥Ìñâ Ï§ëÎã® Í∂åÏû•")
        return 0    # Ïö¥Ìñâ Í∏àÏßÄ

def find_nearest_charging_station(car_position, stations):
    """Í∞ÄÏû• Í∞ÄÍπåÏö¥ Ï∂©Ï†ÑÏÜå Ï∞æÍ∏∞"""
    print(f"üîç ÌòÑÏû¨ ÏúÑÏπò {car_position}ÏóêÏÑú Í∞ÄÏû• Í∞ÄÍπåÏö¥ Ï∂©Ï†ÑÏÜå Í≤ÄÏÉâ:")

    car_x, car_y = car_position
    nearest_station = None
    min_distance = float('inf')

    for station_name, station_pos in stations.items():
        station_x, station_y = station_pos
        distance = abs(car_x - station_x) + abs(car_y - station_y)
        print(f"  {station_name}: Í±∞Î¶¨ {distance}km")

        if distance < min_distance:
            min_distance = distance
            nearest_station = station_name

    print(f"üéØ Ï∂îÏ≤ú: {nearest_station} (Í±∞Î¶¨: {min_distance}km)")
    return nearest_station

def emergency_brake_test():
    """Í∏¥Í∏â Ï†úÎèô ÌÖåÏä§Ìä∏"""
    print("üö® Í∏¥Í∏â Ï†úÎèô ÌÖåÏä§Ìä∏!")

    speeds = [30, 50, 80, 100]  # km/h

    for speed in speeds:
        if speed <= 50:
            brake_distance = speed / 10
            print(f"  ÏÜçÎèÑ {speed}km/h ‚Üí Ï†úÎèôÍ±∞Î¶¨ {brake_distance}m ‚úÖ")
        else:
            brake_distance = speed / 8
            print(f"  ÏÜçÎèÑ {speed}km/h ‚Üí Ï†úÎèôÍ±∞Î¶¨ {brake_distance}m ‚ö†Ô∏è")

# 5Îã®Í≥Ñ ÌÖåÏä§Ìä∏
max_speed = check_weather_condition("ÎπÑ")

charging_stations = {
    "Í∞ïÎÇ®Ï∂©Ï†ÑÏÜå": (5, 3),
    "ÏÑúÏ¥àÏ∂©Ï†ÑÏÜå": (8, 7),
    "ÏÜ°ÌååÏ∂©Ï†ÑÏÜå": (2, 9)
}
find_nearest_charging_station((6, 4), charging_stations)

emergency_brake_test()

print("\n" + "="*60)
print("üèÅ Î™®Îì† Îã®Í≥Ñ ÏôÑÎ£å!")
print("Í∞Å Îã®Í≥ÑÎ≥ÑÎ°ú Î≥µÏÇ¨Ìï¥ÏÑú Í∞úÎ≥Ñ Ïã§ÌñâÌïòÍ±∞ÎÇò")
print("Ï†ÑÏ≤¥Î•º ÌïúÎ≤àÏóê Ïã§ÌñâÌï¥ÏÑú Í≤∞Í≥ºÎ•º ÌôïÏù∏ÌïòÏÑ∏Ïöî!")
print("="*60)
```

```python
# ClassÏôÄ FunctionÏùÑ ÌôúÏö©Ìïú ÏûêÏú®Ï£ºÌñâÏ∞® ÏãúÎÆ¨Î†àÏù¥ÌÑ∞
# ÏΩîÎû©ÏóêÏÑú Î∞îÎ°ú Ïã§Ìñâ Í∞ÄÎä•!

class AutonomousCar:
    """ÏûêÏú®Ï£ºÌñâÏ∞® ÌÅ¥ÎûòÏä§"""

    def __init__(self, car_id, battery=100):
        """ÏûêÏú®Ï£ºÌñâÏ∞® Ï¥àÍ∏∞Ìôî"""
        self.car_id = car_id
        self.battery = battery
        self.position = [0, 0]  # [x, y] Ï¢åÌëú
        self.speed = 0
        self.destination = None
        self.sensors = {
            'camera': True,
            'lidar': True,
            'radar': True,
            'gps': True
        }
        print(f"üöó {self.car_id} ÏûêÏú®Ï£ºÌñâÏ∞® ÏÉùÏÑ± ÏôÑÎ£å!")

    def check_battery(self):
        """Î∞∞ÌÑ∞Î¶¨ ÏÉÅÌÉú ÌôïÏù∏"""
        if self.battery > 50:
            return "Ï∂©Î∂Ñ"
        elif self.battery > 20:
            return "Î≥¥ÌÜµ"
        else:
            return "Î∂ÄÏ°±"

    def move_to(self, x, y):
        """Î™©Ï†ÅÏßÄÎ°ú Ïù¥Îèô"""
        print(f"\nüéØ {self.car_id}: ({x}, {y})Î°ú Ïù¥Îèô ÏãúÏûë")

        start_x, start_y = self.position
        distance = abs(x - start_x) + abs(y - start_y)  # Îß®ÌïòÌÉÑ Í±∞Î¶¨

        # Î∞∞ÌÑ∞Î¶¨ ÏÜåÎ™® Í≥ÑÏÇ∞
        battery_needed = distance * 2

        if self.battery < battery_needed:
            print(f"‚ùå Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°±! ÌòÑÏû¨: {self.battery}%, ÌïÑÏöî: {battery_needed}%")
            return False

        # Ïù¥Îèô ÏãúÎÆ¨Î†àÏù¥ÏÖò
        self.position = [x, y]
        self.battery -= battery_needed

        print(f"‚úÖ ÎèÑÏ∞©! ÌòÑÏû¨ ÏúÑÏπò: {self.position}")
        print(f"üîã ÎÇ®ÏùÄ Î∞∞ÌÑ∞Î¶¨: {self.battery}%")
        return True

    def scan_sensors(self):
        """ÏÑºÏÑú ÏÉÅÌÉú Ïä§Ï∫î"""
        print(f"\nüîç {self.car_id} ÏÑºÏÑú Ïä§Ï∫î Ï§ë...")

        working_sensors = []
        broken_sensors = []

        for sensor_name, is_working in self.sensors.items():
            if is_working:
                working_sensors.append(sensor_name)
            else:
                broken_sensors.append(sensor_name)

        print(f"‚úÖ Ï†ïÏÉÅ ÏÑºÏÑú: {working_sensors}")
        if broken_sensors:
            print(f"‚ùå Í≥†Ïû• ÏÑºÏÑú: {broken_sensors}")

        return len(working_sensors) >= 3  # 3Í∞ú Ïù¥ÏÉÅÏù¥Î©¥ ÏïàÏ†Ñ

    def plan_route(self, waypoints):
        """Í≤ΩÎ°ú Í≥ÑÌöç"""
        print(f"\nüó∫Ô∏è {self.car_id} Í≤ΩÎ°ú Í≥ÑÌöç:")

        total_distance = 0
        current_pos = self.position.copy()

        for i, waypoint in enumerate(waypoints):
            x, y = waypoint
            distance = abs(x - current_pos[0]) + abs(y - current_pos[1])
            total_distance += distance
            current_pos = [x, y]
            print(f"  {i+1}Îã®Í≥Ñ: {waypoint} (Í±∞Î¶¨: {distance})")

        print(f"üìè Ï¥ù Í±∞Î¶¨: {total_distance}")

        # Î∞∞ÌÑ∞Î¶¨Î°ú Í∞à Ïàò ÏûàÎäîÏßÄ ÌôïÏù∏
        battery_needed = total_distance * 2
        if self.battery >= battery_needed:
            print("‚úÖ Î∞∞ÌÑ∞Î¶¨ Ï∂©Î∂Ñ - Í≤ΩÎ°ú Ïã§Ìñâ Í∞ÄÎä•")
            return True
        else:
            print(f"‚ùå Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°± - ÌïÑÏöî: {battery_needed}%, ÌòÑÏû¨: {self.battery}%")
            return False


class TrafficManager:
    """ÍµêÌÜµ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖú ÌÅ¥ÎûòÏä§"""

    def __init__(self):
        """ÍµêÌÜµ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖú Ï¥àÍ∏∞Ìôî"""
        self.intersections = {
            'ÍµêÏ∞®Î°úA': {'signal': 'green', 'time_left': 30},
            'ÍµêÏ∞®Î°úB': {'signal': 'red', 'time_left': 45},
            'ÍµêÏ∞®Î°úC': {'signal': 'yellow', 'time_left': 5}
        }
        self.cars = []
        print("üö¶ ÍµêÌÜµ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖú ÏãúÏûë!")

    def add_car(self, car):
        """Ï∞®Îüâ Îì±Î°ù"""
        self.cars.append(car)
        print(f"üìù {car.car_id} Ï∞®Îüâ Îì±Î°ù ÏôÑÎ£å")

    def check_traffic_lights(self):
        """Ïã†Ìò∏Îì± ÏÉÅÌÉú ÌôïÏù∏"""
        print("\nüö¶ ÌòÑÏû¨ Ïã†Ìò∏Îì± ÏÉÅÌÉú:")

        for intersection, info in self.intersections.items():
            signal = info['signal']
            time_left = info['time_left']

            if signal == 'green':
                status = "üü¢ ÌÜµÍ≥º Í∞ÄÎä•"
            elif signal == 'yellow':
                status = "üü° Ï£ºÏùò ÌïÑÏöî"
            else:
                status = "üî¥ Ï†ïÏßÄ ÌïÑÏöî"

            print(f"  {intersection}: {status} ({time_left}Ï¥à)")

    def find_safe_route(self, start, end):
        """ÏïàÏ†ÑÌïú Í≤ΩÎ°ú Ï∞æÍ∏∞"""
        print(f"\nüõ£Ô∏è {start} ‚Üí {end} ÏïàÏ†Ñ Í≤ΩÎ°ú Í≤ÄÏÉâ")

        # Ïã†Ìò∏Îì± ÏÉÅÌÉú ÌôïÏù∏Ìï¥ÏÑú Í≤ΩÎ°ú Ï∂îÏ≤ú
        green_lights = []
        red_lights = []

        for intersection, info in self.intersections.items():
            if info['signal'] == 'green':
                green_lights.append(intersection)
            elif info['signal'] == 'red':
                red_lights.append(intersection)

        if green_lights:
            recommended = green_lights[0]
            print(f"‚úÖ Ï∂îÏ≤ú Í≤ΩÎ°ú: {recommended} Í≤ΩÏú† (Ïã†Ìò∏ ÏñëÌò∏)")
        else:
            print("‚ö†Ô∏è Î™®Îì† Ïã†Ìò∏Îì± Ï£ºÏùò - ÏÑúÌñâ Ïö¥Ï†Ñ")

    def monitor_fleet(self):
        """Ï∞®Îüâ Ìï®ÎåÄ Î™®ÎãàÌÑ∞ÎßÅ"""
        print("\nüìä Ï∞®Îüâ Ìï®ÎåÄ ÌòÑÌô©:")

        if not self.cars:
            print("Îì±Î°ùÎêú Ï∞®ÎüâÏù¥ ÏóÜÏäµÎãàÎã§.")
            return

        total_cars = len(self.cars)
        low_battery_cars = []
        safe_cars = []

        for car in self.cars:
            battery_status = car.check_battery()
            sensor_ok = car.scan_sensors()

            print(f"\n{car.car_id}:")
            print(f"  ÏúÑÏπò: {car.position}")
            print(f"  Î∞∞ÌÑ∞Î¶¨: {car.battery}% ({battery_status})")
            print(f"  ÏÑºÏÑú ÏÉÅÌÉú: {'Ï†ïÏÉÅ' if sensor_ok else 'Ï†êÍ≤ÄÌïÑÏöî'}")

            if battery_status == "Î∂ÄÏ°±" or not sensor_ok:
                low_battery_cars.append(car.car_id)
            else:
                safe_cars.append(car.car_id)

        print(f"\nüìà Ìï®ÎåÄ ÏöîÏïΩ:")
        print(f"Ï¥ù Ï∞®Îüâ: {total_cars}ÎåÄ")
        print(f"Ï†ïÏÉÅ Ïö¥Ìñâ: {len(safe_cars)}ÎåÄ")
        print(f"Ï†êÍ≤Ä ÌïÑÏöî: {len(low_battery_cars)}ÎåÄ")

        if low_battery_cars:
            print(f"‚ö†Ô∏è Ï†êÍ≤Ä ÎåÄÏÉÅ: {low_battery_cars}")


def simulate_emergency_scenario():
    """ÏùëÍ∏â ÏÉÅÌô© ÏãúÎÆ¨Î†àÏù¥ÏÖò Ìï®Ïàò"""
    print("\nüö® ÏùëÍ∏â ÏÉÅÌô© Î∞úÏÉù!")

    # ÏùëÍ∏âÏ∞®Îüâ ÏÉùÏÑ±
    ambulance = AutonomousCar("ÏùëÍ∏âÏ∞®_001", battery=80)

    # ÏùëÍ∏â Í≤ΩÎ°ú ÏÑ§Ï†ï
    emergency_route = [(5, 0), (5, 5), (10, 5), (10, 10)]

    print("üè• Î≥ëÏõêÏúºÎ°ú Í∏¥Í∏â Ïù¥ÏÜ° Ï§ë...")

    # Í≤ΩÎ°ú Îî∞Îùº Ïù¥Îèô
    for i, waypoint in enumerate(emergency_route):
        x, y = waypoint
        print(f"\nÎã®Í≥Ñ {i+1}: {waypoint}Î°ú Ïù¥Îèô")

        # Ïù¥Îèô Í∞ÄÎä• Ïó¨Î∂Ä ÌôïÏù∏
        if ambulance.move_to(x, y):
            print(f"‚úÖ Îã®Í≥Ñ {i+1} ÏôÑÎ£å")
        else:
            print(f"‚ùå Îã®Í≥Ñ {i+1} Ïã§Ìå® - ÏùëÍ∏â Ï∂©Ï†Ñ ÌïÑÏöî")
            break

    print("üè• Î≥ëÏõê ÎèÑÏ∞©!")


def run_autonomous_driving_simulation():
    """ÏûêÏú®Ï£ºÌñâ ÏãúÎÆ¨Î†àÏù¥ÏÖò Î©îÏù∏ Ìï®Ïàò"""
    print("=" * 50)
    print("üöó ÏûêÏú®Ï£ºÌñâÏ∞® ÏãúÎÆ¨Î†àÏù¥ÏÖò ÏãúÏûë!")
    print("=" * 50)

    # ÍµêÌÜµ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖú ÏÉùÏÑ±
    traffic_manager = TrafficManager()

    # ÏûêÏú®Ï£ºÌñâÏ∞®Îì§ ÏÉùÏÑ±
    car1 = AutonomousCar("ÏûêÏú®Ï∞®_001", battery=90)
    car2 = AutonomousCar("ÏûêÏú®Ï∞®_002", battery=30)
    car3 = AutonomousCar("ÏûêÏú®Ï∞®_003", battery=75)

    # Ï∞®ÎüâÎì§ÏùÑ ÍµêÌÜµ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖúÏóê Îì±Î°ù
    traffic_manager.add_car(car1)
    traffic_manager.add_car(car2)
    traffic_manager.add_car(car3)

    # Ïã†Ìò∏Îì± ÌôïÏù∏
    traffic_manager.check_traffic_lights()

    # ÏïàÏ†Ñ Í≤ΩÎ°ú Í≤ÄÏÉâ
    traffic_manager.find_safe_route("Ï∂úÎ∞úÏßÄ", "Î™©Ï†ÅÏßÄ")

    # Ï∞®ÎüâÎ≥Ñ Í≤ΩÎ°ú Í≥ÑÌöç Î∞è Ïù¥Îèô
    print("\n" + "=" * 30)
    print("üõ£Ô∏è Í∞úÎ≥Ñ Ï∞®Îüâ Ïö¥Ìñâ ÏãúÏûë")
    print("=" * 30)

    # Ï∞®Îüâ 1: Ï†ïÏÉÅ Ïö¥Ìñâ
    route1 = [(3, 3), (6, 6), (10, 10)]
    if car1.plan_route(route1):
        for waypoint in route1:
            car1.move_to(waypoint[0], waypoint[1])

    # Ï∞®Îüâ 2: Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°±ÏúºÎ°ú Ï†úÌïúÏ†Å Ïù¥Îèô
    route2 = [(2, 2), (4, 4)]
    if car2.plan_route(route2):
        for waypoint in route2:
            if not car2.move_to(waypoint[0], waypoint[1]):
                break

    # Ï∞®Îüâ 3: ÏÑºÏÑú Í≥†Ïû• ÏãúÎÆ¨Î†àÏù¥ÏÖò
    car3.sensors['camera'] = False  # Ïπ¥Î©îÎùº Í≥†Ïû•
    car3.sensors['lidar'] = False   # ÎùºÏù¥Îã§ Í≥†Ïû•

    route3 = [(1, 1), (2, 2)]
    if car3.scan_sensors():  # ÏÑºÏÑú ÌôïÏù∏
        car3.plan_route(route3)
    else:
        print(f"‚ö†Ô∏è {car3.car_id}: ÏÑºÏÑú Ïù¥ÏÉÅÏúºÎ°ú Ïö¥Ìñâ Ï§ëÎã®")

    # Ï†ÑÏ≤¥ Ìï®ÎåÄ Î™®ÎãàÌÑ∞ÎßÅ
    traffic_manager.monitor_fleet()

    # ÏùëÍ∏â ÏÉÅÌô© ÏãúÎÆ¨Î†àÏù¥ÏÖò
    simulate_emergency_scenario()

    print("\n" + "=" * 50)
    print("üèÅ ÏãúÎÆ¨Î†àÏù¥ÏÖò ÏôÑÎ£å!")
    print("=" * 50)


# ÏãúÎÆ¨Î†àÏù¥ÏÖò Ïã§Ìñâ
if __name__ == "__main__":
    run_autonomous_driving_simulation()
```
