```python

# ===== 1ë‹¨ê³„: ê¸°ë³¸ ììœ¨ì£¼í–‰ì°¨ í´ë˜ìŠ¤ =====
print("=== 1ë‹¨ê³„: ê¸°ë³¸ ììœ¨ì£¼í–‰ì°¨ ===")

class BasicCar:
    """ê°€ì¥ ê¸°ë³¸ì ì¸ ììœ¨ì£¼í–‰ì°¨"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = 0
        print(f"ğŸš— {self.name} ìƒì„±!")

    def move(self, distance):
        if self.battery >= distance:
            self.battery -= distance
            self.position += distance
            print(f"âœ… {distance}km ì´ë™ ì™„ë£Œ! ìœ„ì¹˜: {self.position}km, ë°°í„°ë¦¬: {self.battery}%")
        else:
            print(f"âŒ ë°°í„°ë¦¬ ë¶€ì¡±! í˜„ì¬: {self.battery}%, í•„ìš”: {distance}%")

    def charge(self, amount):
        self.battery += amount
        if self.battery > 100:
            self.battery = 100
        print(f"ğŸ”‹ ì¶©ì „ ì™„ë£Œ! ë°°í„°ë¦¬: {self.battery}%")

# 1ë‹¨ê³„ í…ŒìŠ¤íŠ¸
car1 = BasicCar("í…ŒìŠ¬ë¼")
car1.move(30)
car1.move(50)
car1.charge(40)
car1.move(50)

print("\n" + "="*50 + "\n")
```

```python
# ===== 2ë‹¨ê³„: ì„¼ì„œ ê¸°ëŠ¥ ì¶”ê°€ =====
print("=== 2ë‹¨ê³„: ì„¼ì„œ ê¸°ëŠ¥ ì¶”ê°€ ===")

class SensorCar:
    """ì„¼ì„œê°€ ìˆëŠ” ììœ¨ì£¼í–‰ì°¨"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = 0
        self.sensors = {
            'camera': True,
            'lidar': True,
            'radar': True
        }
        print(f"ğŸš— {self.name} (ì„¼ì„œ ì¥ì°©) ìƒì„±!")

    def check_sensors(self):
        print(f"ğŸ” {self.name} ì„¼ì„œ ì ê²€:")
        working_count = 0

        for sensor_name, is_working in self.sensors.items():
            status = "ì •ìƒ" if is_working else "ê³ ì¥"
            print(f"  {sensor_name}: {status}")
            if is_working:
                working_count += 1

        if working_count >= 2:
            print("âœ… ì£¼í–‰ ê°€ëŠ¥ (ì„¼ì„œ 2ê°œ ì´ìƒ ì •ìƒ)")
            return True
        else:
            print("âŒ ì£¼í–‰ ë¶ˆê°€ (ì„¼ì„œ ë¶€ì¡±)")
            return False

    def move_safely(self, distance):
        if self.check_sensors():
            if self.battery >= distance:
                self.battery -= distance
                self.position += distance
                print(f"âœ… ì•ˆì „í•˜ê²Œ {distance}km ì´ë™! ìœ„ì¹˜: {self.position}km")
            else:
                print(f"âŒ ë°°í„°ë¦¬ ë¶€ì¡±!")
        else:
            print("âš ï¸ ì„¼ì„œ ì´ìƒìœ¼ë¡œ ì´ë™ ì¤‘ë‹¨!")

# 2ë‹¨ê³„ í…ŒìŠ¤íŠ¸
car2 = SensorCar("í˜„ëŒ€")
car2.move_safely(20)

# ì„¼ì„œ ê³ ì¥ ì‹œë®¬ë ˆì´ì…˜
car2.sensors['camera'] = False
car2.sensors['lidar'] = False
car2.move_safely(20)

print("\n" + "="*50 + "\n")
```

```python
# ===== 3ë‹¨ê³„: ì‹ í˜¸ë“± ì‹œìŠ¤í…œ ì¶”ê°€ =====
print("=== 3ë‹¨ê³„: ì‹ í˜¸ë“± ì‹œìŠ¤í…œ ===")

class TrafficLight:
    """ì‹ í˜¸ë“± í´ë˜ìŠ¤"""

    def __init__(self, name, initial_color='red'):
        self.name = name
        self.color = initial_color
        self.time_left = 30
        print(f"ğŸš¦ {self.name} ì‹ í˜¸ë“± ì„¤ì¹˜ ({self.color})")

    def get_status(self):
        if self.color == 'green':
            return True  # í†µê³¼ ê°€ëŠ¥
        else:
            return False  # ì •ì§€

    def change_signal(self):
        if self.color == 'red':
            self.color = 'green'
        elif self.color == 'green':
            self.color = 'yellow'
        else:
            self.color = 'red'
        print(f"ğŸš¦ {self.name}: {self.color} ì‹ í˜¸ë¡œ ë³€ê²½")

class SmartCar:
    """ì‹ í˜¸ë“±ì„ ì¸ì‹í•˜ëŠ” ììœ¨ì£¼í–‰ì°¨"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = 0
        print(f"ğŸš— {self.name} (ìŠ¤ë§ˆíŠ¸) ìƒì„±!")

    def approach_intersection(self, traffic_light):
        print(f"\nğŸš— {self.name}: {traffic_light.name} ì ‘ê·¼")
        print(f"ğŸš¦ í˜„ì¬ ì‹ í˜¸: {traffic_light.color}")

        if traffic_light.get_status():
            print("âœ… ì´ˆë¡ë¶ˆ - í†µê³¼!")
            self.position += 10
            self.battery -= 5
        else:
            print("ğŸ›‘ ë¹¨ê°„ë¶ˆ/ë…¸ë€ë¶ˆ - ì •ì§€!")
            print("â³ ì‹ í˜¸ ë³€ê²½ ëŒ€ê¸° ì¤‘...")

# 3ë‹¨ê³„ í…ŒìŠ¤íŠ¸
light1 = TrafficLight("êµì°¨ë¡œA", 'red')
car3 = SmartCar("BMW")

car3.approach_intersection(light1)
light1.change_signal()  # ì´ˆë¡ë¶ˆë¡œ ë³€ê²½
car3.approach_intersection(light1)

print("\n" + "="*50 + "\n")
```

```python
# ===== 4ë‹¨ê³„: ê²½ë¡œ ê³„íš ê¸°ëŠ¥ =====
print("=== 4ë‹¨ê³„: ê²½ë¡œ ê³„íš ===")

class NavigationCar:
    """ê²½ë¡œ ê³„íšì´ ê°€ëŠ¥í•œ ììœ¨ì£¼í–‰ì°¨"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = [0, 0]  # [x, y] ì¢Œí‘œ
        print(f"ğŸš— {self.name} (ë‚´ë¹„ê²Œì´ì…˜) ìƒì„±!")

    def plan_route(self, waypoints):
        print(f"\nğŸ—ºï¸ {self.name} ê²½ë¡œ ê³„íš:")
        total_distance = 0

        current_x, current_y = self.position

        for i, waypoint in enumerate(waypoints):
            target_x, target_y = waypoint
            distance = abs(target_x - current_x) + abs(target_y - current_y)
            total_distance += distance
            print(f"  {i+1}ë‹¨ê³„: ({target_x}, {target_y}) - ê±°ë¦¬ {distance}km")
            current_x, current_y = target_x, target_y

        print(f"ğŸ“ ì´ ê±°ë¦¬: {total_distance}km")
        battery_needed = total_distance * 2

        if self.battery >= battery_needed:
            print("âœ… ë°°í„°ë¦¬ ì¶©ë¶„ - ì¶œë°œ ê°€ëŠ¥!")
            return True
        else:
            print(f"âŒ ë°°í„°ë¦¬ ë¶€ì¡± - í•„ìš”: {battery_needed}%, í˜„ì¬: {self.battery}%")
            return False

    def move_to_waypoint(self, x, y):
        start_x, start_y = self.position
        distance = abs(x - start_x) + abs(y - start_y)
        battery_cost = distance * 2

        if self.battery >= battery_cost:
            self.position = [x, y]
            self.battery -= battery_cost
            print(f"âœ… ({x}, {y}) ë„ì°©! ë°°í„°ë¦¬: {self.battery}%")
            return True
        else:
            print(f"âŒ ë°°í„°ë¦¬ ë¶€ì¡±ìœ¼ë¡œ ì´ë™ ì‹¤íŒ¨!")
            return False

# 4ë‹¨ê³„ í…ŒìŠ¤íŠ¸
car4 = NavigationCar("êµ¬ê¸€ì¹´")
route = [(3, 0), (3, 5), (8, 5), (8, 10)]

if car4.plan_route(route):
    print("\nğŸ›£ï¸ ê²½ë¡œ ì‹¤í–‰:")
    for waypoint in route:
        x, y = waypoint
        if not car4.move_to_waypoint(x, y):
            break

print("\n" + "="*50 + "\n")
```

```python
# Classì™€ Functionì„ í™œìš©í•œ ììœ¨ì£¼í–‰ì°¨ ì‹œë®¬ë ˆì´í„°
# ì½”ë©ì—ì„œ ë°”ë¡œ ì‹¤í–‰ ê°€ëŠ¥!

class AutonomousCar:
    """ììœ¨ì£¼í–‰ì°¨ í´ë˜ìŠ¤"""

    def __init__(self, car_id, battery=100):
        """ììœ¨ì£¼í–‰ì°¨ ì´ˆê¸°í™”"""
        self.car_id = car_id
        self.battery = battery
        self.position = [0, 0]  # [x, y] ì¢Œí‘œ
        self.speed = 0
        self.destination = None
        self.sensors = {
            'camera': True,
            'lidar': True,
            'radar': True,
            'gps': True
        }
        print(f"ğŸš— {self.car_id} ììœ¨ì£¼í–‰ì°¨ ìƒì„± ì™„ë£Œ!")

    def check_battery(self):
        """ë°°í„°ë¦¬ ìƒíƒœ í™•ì¸"""
        if self.battery > 50:
            return "ì¶©ë¶„"
        elif self.battery > 20:
            return "ë³´í†µ"
        else:
            return "ë¶€ì¡±"

    def move_to(self, x, y):
        """ëª©ì ì§€ë¡œ ì´ë™"""
        print(f"\nğŸ¯ {self.car_id}: ({x}, {y})ë¡œ ì´ë™ ì‹œì‘")

        start_x, start_y = self.position
        distance = abs(x - start_x) + abs(y - start_y)  # ë§¨í•˜íƒ„ ê±°ë¦¬

        # ë°°í„°ë¦¬ ì†Œëª¨ ê³„ì‚°
        battery_needed = distance * 2

        if self.battery < battery_needed:
            print(f"âŒ ë°°í„°ë¦¬ ë¶€ì¡±! í˜„ì¬: {self.battery}%, í•„ìš”: {battery_needed}%")
            return False

        # ì´ë™ ì‹œë®¬ë ˆì´ì…˜
        self.position = [x, y]
        self.battery -= battery_needed

        print(f"âœ… ë„ì°©! í˜„ì¬ ìœ„ì¹˜: {self.position}")
        print(f"ğŸ”‹ ë‚¨ì€ ë°°í„°ë¦¬: {self.battery}%")
        return True

    def scan_sensors(self):
        """ì„¼ì„œ ìƒíƒœ ìŠ¤ìº”"""
        print(f"\nğŸ” {self.car_id} ì„¼ì„œ ìŠ¤ìº” ì¤‘...")

        working_sensors = []
        broken_sensors = []

        for sensor_name, is_working in self.sensors.items():
            if is_working:
                working_sensors.append(sensor_name)
            else:
                broken_sensors.append(sensor_name)

        print(f"âœ… ì •ìƒ ì„¼ì„œ: {working_sensors}")
        if broken_sensors:
            print(f"âŒ ê³ ì¥ ì„¼ì„œ: {broken_sensors}")

        return len(working_sensors) >= 3  # 3ê°œ ì´ìƒì´ë©´ ì•ˆì „

    def plan_route(self, waypoints):
        """ê²½ë¡œ ê³„íš"""
        print(f"\nğŸ—ºï¸ {self.car_id} ê²½ë¡œ ê³„íš:")

        total_distance = 0
        current_pos = self.position.copy()

        for i, waypoint in enumerate(waypoints):
            x, y = waypoint
            distance = abs(x - current_pos[0]) + abs(y - current_pos[1])
            total_distance += distance
            current_pos = [x, y]
            print(f"  {i+1}ë‹¨ê³„: {waypoint} (ê±°ë¦¬: {distance})")

        print(f"ğŸ“ ì´ ê±°ë¦¬: {total_distance}")

        # ë°°í„°ë¦¬ë¡œ ê°ˆ ìˆ˜ ìˆëŠ”ì§€ í™•ì¸
        battery_needed = total_distance * 2
        if self.battery >= battery_needed:
            print("âœ… ë°°í„°ë¦¬ ì¶©ë¶„ - ê²½ë¡œ ì‹¤í–‰ ê°€ëŠ¥")
            return True
        else:
            print(f"âŒ ë°°í„°ë¦¬ ë¶€ì¡± - í•„ìš”: {battery_needed}%, í˜„ì¬: {self.battery}%")
            return False


class TrafficManager:
    """êµí†µ ê´€ë¦¬ ì‹œìŠ¤í…œ í´ë˜ìŠ¤"""

    def __init__(self):
        """êµí†µ ê´€ë¦¬ ì‹œìŠ¤í…œ ì´ˆê¸°í™”"""
        self.intersections = {
            'êµì°¨ë¡œA': {'signal': 'green', 'time_left': 30},
            'êµì°¨ë¡œB': {'signal': 'red', 'time_left': 45},
            'êµì°¨ë¡œC': {'signal': 'yellow', 'time_left': 5}
        }
        self.cars = []
        print("ğŸš¦ êµí†µ ê´€ë¦¬ ì‹œìŠ¤í…œ ì‹œì‘!")

    def add_car(self, car):
        """ì°¨ëŸ‰ ë“±ë¡"""
        self.cars.append(car)
        print(f"ğŸ“ {car.car_id} ì°¨ëŸ‰ ë“±ë¡ ì™„ë£Œ")

    def check_traffic_lights(self):
        """ì‹ í˜¸ë“± ìƒíƒœ í™•ì¸"""
        print("\nğŸš¦ í˜„ì¬ ì‹ í˜¸ë“± ìƒíƒœ:")

        for intersection, info in self.intersections.items():
            signal = info['signal']
            time_left = info['time_left']

            if signal == 'green':
                status = "ğŸŸ¢ í†µê³¼ ê°€ëŠ¥"
            elif signal == 'yellow':
                status = "ğŸŸ¡ ì£¼ì˜ í•„ìš”"
            else:
                status = "ğŸ”´ ì •ì§€ í•„ìš”"

            print(f"  {intersection}: {status} ({time_left}ì´ˆ)")

    def find_safe_route(self, start, end):
        """ì•ˆì „í•œ ê²½ë¡œ ì°¾ê¸°"""
        print(f"\nğŸ›£ï¸ {start} â†’ {end} ì•ˆì „ ê²½ë¡œ ê²€ìƒ‰")

        # ì‹ í˜¸ë“± ìƒíƒœ í™•ì¸í•´ì„œ ê²½ë¡œ ì¶”ì²œ
        green_lights = []
        red_lights = []

        for intersection, info in self.intersections.items():
            if info['signal'] == 'green':
                green_lights.append(intersection)
            elif info['signal'] == 'red':
                red_lights.append(intersection)

        if green_lights:
            recommended = green_lights[0]
            print(f"âœ… ì¶”ì²œ ê²½ë¡œ: {recommended} ê²½ìœ  (ì‹ í˜¸ ì–‘í˜¸)")
        else:
            print("âš ï¸ ëª¨ë“  ì‹ í˜¸ë“± ì£¼ì˜ - ì„œí–‰ ìš´ì „")

    def monitor_fleet(self):
        """ì°¨ëŸ‰ í•¨ëŒ€ ëª¨ë‹ˆí„°ë§"""
        print("\nğŸ“Š ì°¨ëŸ‰ í•¨ëŒ€ í˜„í™©:")

        if not self.cars:
            print("ë“±ë¡ëœ ì°¨ëŸ‰ì´ ì—†ìŠµë‹ˆë‹¤.")
            return

        total_cars = len(self.cars)
        low_battery_cars = []
        safe_cars = []

        for car in self.cars:
            battery_status = car.check_battery()
            sensor_ok = car.scan_sensors()

            print(f"\n{car.car_id}:")
            print(f"  ìœ„ì¹˜: {car.position}")
            print(f"  ë°°í„°ë¦¬: {car.battery}% ({battery_status})")
            print(f"  ì„¼ì„œ ìƒíƒœ: {'ì •ìƒ' if sensor_ok else 'ì ê²€í•„ìš”'}")

            if battery_status == "ë¶€ì¡±" or not sensor_ok:
                low_battery_cars.append(car.car_id)
            else:
                safe_cars.append(car.car_id)

        print(f"\nğŸ“ˆ í•¨ëŒ€ ìš”ì•½:")
        print(f"ì´ ì°¨ëŸ‰: {total_cars}ëŒ€")
        print(f"ì •ìƒ ìš´í–‰: {len(safe_cars)}ëŒ€")
        print(f"ì ê²€ í•„ìš”: {len(low_battery_cars)}ëŒ€")

        if low_battery_cars:
            print(f"âš ï¸ ì ê²€ ëŒ€ìƒ: {low_battery_cars}")


def simulate_emergency_scenario():
    """ì‘ê¸‰ ìƒí™© ì‹œë®¬ë ˆì´ì…˜ í•¨ìˆ˜"""
    print("\nğŸš¨ ì‘ê¸‰ ìƒí™© ë°œìƒ!")

    # ì‘ê¸‰ì°¨ëŸ‰ ìƒì„±
    ambulance = AutonomousCar("ì‘ê¸‰ì°¨_001", battery=80)

    # ì‘ê¸‰ ê²½ë¡œ ì„¤ì •
    emergency_route = [(5, 0), (5, 5), (10, 5), (10, 10)]

    print("ğŸ¥ ë³‘ì›ìœ¼ë¡œ ê¸´ê¸‰ ì´ì†¡ ì¤‘...")

    # ê²½ë¡œ ë”°ë¼ ì´ë™
    for i, waypoint in enumerate(emergency_route):
        x, y = waypoint
        print(f"\në‹¨ê³„ {i+1}: {waypoint}ë¡œ ì´ë™")

        # ì´ë™ ê°€ëŠ¥ ì—¬ë¶€ í™•ì¸
        if ambulance.move_to(x, y):
            print(f"âœ… ë‹¨ê³„ {i+1} ì™„ë£Œ")
        else:
            print(f"âŒ ë‹¨ê³„ {i+1} ì‹¤íŒ¨ - ì‘ê¸‰ ì¶©ì „ í•„ìš”")
            break

    print("ğŸ¥ ë³‘ì› ë„ì°©!")


def run_autonomous_driving_simulation():
    """ììœ¨ì£¼í–‰ ì‹œë®¬ë ˆì´ì…˜ ë©”ì¸ í•¨ìˆ˜"""
    print("=" * 50)
    print("ğŸš— ììœ¨ì£¼í–‰ì°¨ ì‹œë®¬ë ˆì´ì…˜ ì‹œì‘!")
    print("=" * 50)

    # êµí†µ ê´€ë¦¬ ì‹œìŠ¤í…œ ìƒì„±
    traffic_manager = TrafficManager()

    # ììœ¨ì£¼í–‰ì°¨ë“¤ ìƒì„±
    car1 = AutonomousCar("ììœ¨ì°¨_001", battery=90)
    car2 = AutonomousCar("ììœ¨ì°¨_002", battery=30)
    car3 = AutonomousCar("ììœ¨ì°¨_003", battery=75)

    # ì°¨ëŸ‰ë“¤ì„ êµí†µ ê´€ë¦¬ ì‹œìŠ¤í…œì— ë“±ë¡
    traffic_manager.add_car(car1)
    traffic_manager.add_car(car2)
    traffic_manager.add_car(car3)

    # ì‹ í˜¸ë“± í™•ì¸
    traffic_manager.check_traffic_lights()

    # ì•ˆì „ ê²½ë¡œ ê²€ìƒ‰
    traffic_manager.find_safe_route("ì¶œë°œì§€", "ëª©ì ì§€")

    # ì°¨ëŸ‰ë³„ ê²½ë¡œ ê³„íš ë° ì´ë™
    print("\n" + "=" * 30)
    print("ğŸ›£ï¸ ê°œë³„ ì°¨ëŸ‰ ìš´í–‰ ì‹œì‘")
    print("=" * 30)

    # ì°¨ëŸ‰ 1: ì •ìƒ ìš´í–‰
    route1 = [(3, 3), (6, 6), (10, 10)]
    if car1.plan_route(route1):
        for waypoint in route1:
            car1.move_to(waypoint[0], waypoint[1])

    # ì°¨ëŸ‰ 2: ë°°í„°ë¦¬ ë¶€ì¡±ìœ¼ë¡œ ì œí•œì  ì´ë™
    route2 = [(2, 2), (4, 4)]
    if car2.plan_route(route2):
        for waypoint in route2:
            if not car2.move_to(waypoint[0], waypoint[1]):
                break

    # ì°¨ëŸ‰ 3: ì„¼ì„œ ê³ ì¥ ì‹œë®¬ë ˆì´ì…˜
    car3.sensors['camera'] = False  # ì¹´ë©”ë¼ ê³ ì¥
    car3.sensors['lidar'] = False   # ë¼ì´ë‹¤ ê³ ì¥

    route3 = [(1, 1), (2, 2)]
    if car3.scan_sensors():  # ì„¼ì„œ í™•ì¸
        car3.plan_route(route3)
    else:
        print(f"âš ï¸ {car3.car_id}: ì„¼ì„œ ì´ìƒìœ¼ë¡œ ìš´í–‰ ì¤‘ë‹¨")

    # ì „ì²´ í•¨ëŒ€ ëª¨ë‹ˆí„°ë§
    traffic_manager.monitor_fleet()

    # ì‘ê¸‰ ìƒí™© ì‹œë®¬ë ˆì´ì…˜
    simulate_emergency_scenario()

    print("\n" + "=" * 50)
    print("ğŸ ì‹œë®¬ë ˆì´ì…˜ ì™„ë£Œ!")
    print("=" * 50)


# ì‹œë®¬ë ˆì´ì…˜ ì‹¤í–‰
if __name__ == "__main__":
    run_autonomous_driving_simulation()
```

```python
# __init__ ë©”ì„œë“œ ì™„ì „ ê°€ì´ë“œ with ììœ¨ì£¼í–‰ ì¢…í•© ì˜ˆì œ

# ===== __init__ ê¸°ë³¸ ì´í•´ =====
print("=== __init__ ë©”ì„œë“œë€? ===")
print("ğŸ”§ __init__ì€ 'ìƒì„±ì'ë¼ê³  ë¶ˆë¦¬ëŠ” íŠ¹ë³„í•œ ë©”ì„œë“œì…ë‹ˆë‹¤")
print("ğŸ¯ í´ë˜ìŠ¤ë¡œë¶€í„° ê°ì²´ë¥¼ ìƒì„±í•  ë•Œ 'ìë™ìœ¼ë¡œ' ì‹¤í–‰ë©ë‹ˆë‹¤")
print("ğŸ“ ê°ì²´ì˜ 'ì´ˆê¸° ìƒíƒœ'ë¥¼ ì„¤ì •í•˜ëŠ” ì—­í• ì„ í•©ë‹ˆë‹¤")
print()

# ===== 1ë‹¨ê³„: ê°€ì¥ ê¸°ë³¸ì ì¸ __init__ =====
print("=== 1ë‹¨ê³„: ê¸°ë³¸ __init__ ===")

class SimpleCar:
    def __init__(self):  # ë§¤ê°œë³€ìˆ˜ ì—†ëŠ” ê¸°ë³¸í˜•
        print("ğŸš— SimpleCar __init__ ì‹¤í–‰ë¨!")
        self.speed = 0
        self.engine = "off"
        print(f"   ì´ˆê¸° ì†ë„: {self.speed}")
        print(f"   ì´ˆê¸° ì—”ì§„ ìƒíƒœ: {self.engine}")

# ê°ì²´ ìƒì„± - __init__ì´ ìë™ ì‹¤í–‰ë¨
print("car1 = SimpleCar() ì‹¤í–‰:")
car1 = SimpleCar()
print()

# ===== 2ë‹¨ê³„: ë§¤ê°œë³€ìˆ˜ê°€ ìˆëŠ” __init__ =====
print("=== 2ë‹¨ê³„: ë§¤ê°œë³€ìˆ˜ ë°›ëŠ” __init__ ===")

class BasicAutoCar:
    def __init__(self, car_name, initial_battery):  # ë§¤ê°œë³€ìˆ˜ ë°›ê¸°
        print(f"ğŸš— {car_name} ìë™ì°¨ ìƒì„± ì¤‘...")

        # self.ì†ì„±ëª… = ê°’ ìœ¼ë¡œ ê°ì²´ì˜ ì†ì„± ì„¤ì •
        self.name = car_name
        self.battery = initial_battery
        self.position = [0, 0]  # ë¦¬ìŠ¤íŠ¸ ì‚¬ìš©
        self.sensors = {        # ë”•ì…”ë„ˆë¦¬ ì‚¬ìš©
            'camera': True,
            'lidar': True,
            'radar': False
        }

        print(f"   ì´ë¦„: {self.name}")
        print(f"   ë°°í„°ë¦¬: {self.battery}%")
        print(f"   ìœ„ì¹˜: {self.position}")
        print(f"   ì„¼ì„œ ìƒíƒœ: {self.sensors}")
        print("âœ… ì´ˆê¸°í™” ì™„ë£Œ!")

# ë§¤ê°œë³€ìˆ˜ë¥¼ ì£¼ë©´ì„œ ê°ì²´ ìƒì„±
print("car2 = BasicAutoCar('í…ŒìŠ¬ë¼', 85) ì‹¤í–‰:")
car2 = BasicAutoCar('í…ŒìŠ¬ë¼', 85)
print()

print("car3 = BasicAutoCar('í˜„ëŒ€', 60) ì‹¤í–‰:")
car3 = BasicAutoCar('í˜„ëŒ€', 60)
print()

# ===== 3ë‹¨ê³„: ë³µí•© ë°ì´í„°ë¥¼ ë°›ëŠ” __init__ =====
print("=== 3ë‹¨ê³„: ë³µí•© ë°ì´í„° ë°›ëŠ” __init__ ===")

class AdvancedAutoCar:
    def __init__(self, car_info, sensor_list, route_data):
        """
        car_info: íŠœí”Œ (ì´ë¦„, ë°°í„°ë¦¬, íƒ€ì…)
        sensor_list: ë¦¬ìŠ¤íŠ¸ ['camera', 'lidar', 'radar']
        route_data: ë”•ì…”ë„ˆë¦¬ {'start': (0,0), 'end': (10,10)}
        """
        print("ğŸš— ê³ ê¸‰ ììœ¨ì£¼í–‰ì°¨ ìƒì„± ì¤‘...")

        # íŠœí”Œ ì–¸íŒ¨í‚¹ìœ¼ë¡œ ì •ë³´ ë¶„ë¦¬
        name, battery, car_type = car_info
        self.name = name
        self.battery = battery
        self.car_type = car_type

        # ë¦¬ìŠ¤íŠ¸ë¥¼ ë”•ì…”ë„ˆë¦¬ë¡œ ë³€í™˜ (ëª¨ë“  ì„¼ì„œë¥¼ Trueë¡œ ì„¤ì •)
        self.sensors = {}
        for sensor in sensor_list:  # forë¬¸ ì‚¬ìš©
            self.sensors[sensor] = True

        # ë”•ì…”ë„ˆë¦¬ ì •ë³´ ì €ì¥
        self.start_position = route_data['start']
        self.end_position = route_data['end']
        self.current_position = list(self.start_position)  # ë¦¬ìŠ¤íŠ¸ë¡œ ë³µì‚¬

        # ì¶”ê°€ ê³„ì‚°
        start_x, start_y = self.start_position  # íŠœí”Œ ì–¸íŒ¨í‚¹
        end_x, end_y = self.end_position
        self.total_distance = abs(end_x - start_x) + abs(end_y - start_y)

        print(f"   ì°¨ëŸ‰ëª…: {self.name} ({self.car_type})")
        print(f"   ë°°í„°ë¦¬: {self.battery}%")
        print(f"   ì„¼ì„œ: {list(self.sensors.keys())}")
        print(f"   ì¶œë°œì§€: {self.start_position}")
        print(f"   ëª©ì ì§€: {self.end_position}")
        print(f"   ì´ ê±°ë¦¬: {self.total_distance}km")
        print("âœ… ê³ ê¸‰ ì´ˆê¸°í™” ì™„ë£Œ!")

    def move_to_destination(self):
        """ëª©ì ì§€ë¡œ ì´ë™í•˜ëŠ” ë©”ì„œë“œ"""
        print(f"\nğŸ›£ï¸ {self.name} ëª©ì ì§€ë¡œ ì´ë™ ì‹œì‘!")

        # ifë¬¸ìœ¼ë¡œ ë°°í„°ë¦¬ í™•ì¸
        battery_needed = self.total_distance * 2
        if self.battery >= battery_needed:
            self.current_position = list(self.end_position)
            self.battery -= battery_needed
            print(f"âœ… ëª©ì ì§€ ë„ì°©! í˜„ì¬ ìœ„ì¹˜: {self.current_position}")
            print(f"ğŸ”‹ ë‚¨ì€ ë°°í„°ë¦¬: {self.battery}%")
        else:
            print(f"âŒ ë°°í„°ë¦¬ ë¶€ì¡±! í•„ìš”: {battery_needed}%, í˜„ì¬: {self.battery}%")

    def check_sensors(self):
        """ì„¼ì„œ ìƒíƒœ ì²´í¬"""
        print(f"\nğŸ” {self.name} ì„¼ì„œ ì ê²€:")
        working_sensors = []

        for sensor_name, is_working in self.sensors.items():  # ë”•ì…”ë„ˆë¦¬ ìˆœíšŒ
            if is_working:  # ifë¬¸ìœ¼ë¡œ ìƒíƒœ í™•ì¸
                working_sensors.append(sensor_name)
                print(f"  âœ… {sensor_name}: ì •ìƒ")
            else:
                print(f"  âŒ {sensor_name}: ê³ ì¥")

        return len(working_sensors) >= 2  # 2ê°œ ì´ìƒì´ë©´ ì•ˆì „

# ë³µí•© ë°ì´í„°ë¡œ ê°ì²´ ìƒì„±
car_info = ('BMW X5', 90, 'SUV')  # íŠœí”Œ
sensor_list = ['camera', 'lidar', 'radar', 'ultrasonic']  # ë¦¬ìŠ¤íŠ¸
route_data = {'start': (0, 0), 'end': (15, 20)}  # ë”•ì…”ë„ˆë¦¬

print("ê³ ê¸‰ ììœ¨ì£¼í–‰ì°¨ ìƒì„±:")
advanced_car = AdvancedAutoCar(car_info, sensor_list, route_data)

# ë©”ì„œë“œ ì‚¬ìš©
advanced_car.check_sensors()
advanced_car.move_to_destination()
print()

# ===== 4ë‹¨ê³„: ì¡°ê±´ë¶€ ì´ˆê¸°í™”ê°€ ìˆëŠ” __init__ =====
print("=== 4ë‹¨ê³„: ì¡°ê±´ë¶€ ì´ˆê¸°í™” ===")

class SmartAutoCar:
    def __init__(self, name, car_class='economy', weather='clear'):
        """
        ì°¨ëŸ‰ ë“±ê¸‰ê³¼ ë‚ ì”¨ì— ë”°ë¼ ë‹¤ë¥´ê²Œ ì´ˆê¸°í™”
        """
        print(f"ğŸš— {name} ìŠ¤ë§ˆíŠ¸ì¹´ ìƒì„± ì¤‘... (ë“±ê¸‰: {car_class}, ë‚ ì”¨: {weather})")

        self.name = name
        self.weather = weather

        # ifë¬¸ìœ¼ë¡œ ì°¨ëŸ‰ ë“±ê¸‰ì— ë”°ë¥¸ ë‹¤ë¥¸ ì´ˆê¸°í™”
        if car_class == 'economy':
            self.battery = 60
            self.max_speed = 80
            self.sensors = ['camera', 'basic_radar']
        elif car_class == 'premium':
            self.battery = 90
            self.max_speed = 120
            self.sensors = ['camera', 'lidar', 'radar', 'ultrasonic']
        elif car_class == 'luxury':
            self.battery = 100
            self.max_speed = 150
            self.sensors = ['camera', 'lidar', 'radar', 'ultrasonic', 'thermal']
        else:
            # ê¸°ë³¸ê°’ ì„¤ì •
            self.battery = 50
            self.max_speed = 60
            self.sensors = ['camera']

        # ë‚ ì”¨ì— ë”°ë¥¸ ì¶”ê°€ ì„¤ì •
        if weather == 'rain':
            self.max_speed = int(self.max_speed * 0.7)  # 30% ê°ì†
            self.safety_distance = 2.0
        elif weather == 'snow':
            self.max_speed = int(self.max_speed * 0.5)  # 50% ê°ì†
            self.safety_distance = 3.0
        else:  # clear weather
            self.safety_distance = 1.0

        # ì„¼ì„œë¥¼ ë”•ì…”ë„ˆë¦¬ë¡œ ë³€í™˜
        self.sensor_status = {}
        for sensor in self.sensors:  # forë¬¸ìœ¼ë¡œ ì„¼ì„œ ë”•ì…”ë„ˆë¦¬ ìƒì„±
            self.sensor_status[sensor] = True

        print(f"   ë°°í„°ë¦¬: {self.battery}%")
        print(f"   ìµœëŒ€ì†ë„: {self.max_speed}km/h")
        print(f"   ì•ˆì „ê±°ë¦¬: {self.safety_distance}ë°°")
        print(f"   ì„¼ì„œ: {self.sensors}")
        print("âœ… ìŠ¤ë§ˆíŠ¸ ì´ˆê¸°í™” ì™„ë£Œ!")

    def get_driving_recommendation(self):
        """ì£¼í–‰ ê¶Œì¥ì‚¬í•­ ì œê³µ"""
        print(f"\nğŸ“‹ {self.name} ì£¼í–‰ ê¶Œì¥ì‚¬í•­:")
        print(f"í˜„ì¬ ë‚ ì”¨: {self.weather}")
        print(f"ê¶Œì¥ ìµœëŒ€ì†ë„: {self.max_speed}km/h")
        print(f"ì•ˆì „ê±°ë¦¬ ë°°ìˆ˜: {self.safety_distance}")

        # ì„¼ì„œ ìƒíƒœ í™•ì¸
        active_sensors = len([s for s in self.sensor_status.values() if s])
        print(f"í™œì„± ì„¼ì„œ: {active_sensors}ê°œ")

        # ifë¬¸ìœ¼ë¡œ ì¢…í•© ê¶Œì¥ì‚¬í•­
        if self.weather == 'clear' and active_sensors >= 3:
            print("âœ… ì •ìƒ ìš´í–‰ ê°€ëŠ¥")
        elif self.weather in ['rain', 'snow']:
            print("âš ï¸ ì„œí–‰ ìš´ì „ í•„ìˆ˜")
        else:
            print("ğŸ”¶ ì£¼ì˜ ìš´ì „ í•„ìš”")

# ë‹¤ì–‘í•œ ì¡°ê±´ìœ¼ë¡œ ê°ì²´ ìƒì„±
print("economy ë“±ê¸‰, ë§‘ì€ ë‚ ì”¨:")
car_eco = SmartAutoCar("ê²½ì œí˜•ì¹´", 'economy', 'clear')
car_eco.get_driving_recommendation()
print()

print("luxury ë“±ê¸‰, ë¹„ì˜¤ëŠ” ë‚ ì”¨:")
car_luxury = SmartAutoCar("ëŸ­ì…”ë¦¬ì¹´", 'luxury', 'rain')
car_luxury.get_driving_recommendation()
print()

# ===== 5ë‹¨ê³„: ì¢…í•© ì‘ìš© - ììœ¨ì£¼í–‰ í•¨ëŒ€ ê´€ë¦¬ ì‹œìŠ¤í…œ =====
print("=== 5ë‹¨ê³„: ì¢…í•© ì‘ìš© - í•¨ëŒ€ ê´€ë¦¬ ì‹œìŠ¤í…œ ===")

class FleetManager:
    """ììœ¨ì£¼í–‰ í•¨ëŒ€ ê´€ë¦¬ ì‹œìŠ¤í…œ"""

    def __init__(self, company_name, fleet_data):
        """
        fleet_data: ë¦¬ìŠ¤íŠ¸ [
            {'name': 'Car1', 'type': 'sedan', 'battery': 80, 'route': [(0,0), (10,5)]},
            {'name': 'Car2', 'type': 'suv', 'battery': 60, 'route': [(5,5), (15,10)]}
        ]
        """
        print(f"ğŸ¢ {company_name} í•¨ëŒ€ ê´€ë¦¬ ì‹œìŠ¤í…œ ì´ˆê¸°í™” ì¤‘...")

        self.company = company_name
        self.vehicles = []  # ì°¨ëŸ‰ ê°ì²´ë“¤ì„ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
        self.total_vehicles = len(fleet_data)

        # forë¬¸ìœ¼ë¡œ ê° ì°¨ëŸ‰ ë°ì´í„°ë¥¼ ì²˜ë¦¬
        for i, vehicle_data in enumerate(fleet_data):
            print(f"\nì°¨ëŸ‰ {i+1}/{self.total_vehicles} ë“±ë¡ ì¤‘:")

            # ë”•ì…”ë„ˆë¦¬ì—ì„œ ë°ì´í„° ì¶”ì¶œ
            name = vehicle_data['name']
            car_type = vehicle_data['type']
            battery = vehicle_data['battery']
            route = vehicle_data['route']

            # ê²½ë¡œ ì •ë³´ ì²˜ë¦¬ (íŠœí”Œ ì–¸íŒ¨í‚¹)
            start_point, end_point = route[0], route[-1]

            # ì°¨ëŸ‰ ì •ë³´ë¥¼ íŠœí”Œë¡œ êµ¬ì„±
            car_info = (name, battery, car_type)

            # ê²½ë¡œ ì •ë³´ë¥¼ ë”•ì…”ë„ˆë¦¬ë¡œ êµ¬ì„±
            route_info = {'start': start_point, 'end': end_point}

            # ì°¨ëŸ‰ íƒ€ì…ì— ë”°ë¥¸ ì„¼ì„œ ì„¤ì •
            if car_type == 'sedan':
                sensors = ['camera', 'radar']
            elif car_type == 'suv':
                sensors = ['camera', 'lidar', 'radar']
            else:  # truck ë“±
                sensors = ['camera', 'lidar', 'radar', 'ultrasonic']

            # AdvancedAutoCar ê°ì²´ ìƒì„±
            vehicle = AdvancedAutoCar(car_info, sensors, route_info)
            self.vehicles.append(vehicle)  # ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€

        print(f"\nâœ… {self.company} í•¨ëŒ€ ì´ˆê¸°í™” ì™„ë£Œ!")
        print(f"ì´ ë“±ë¡ ì°¨ëŸ‰: {self.total_vehicles}ëŒ€")

    def start_fleet_operation(self):
        """í•¨ëŒ€ ìš´í–‰ ì‹œì‘"""
        print(f"\nğŸš€ {self.company} í•¨ëŒ€ ìš´í–‰ ì‹œì‘!")

        for vehicle in self.vehicles:  # forë¬¸ìœ¼ë¡œ ëª¨ë“  ì°¨ëŸ‰ ì²˜ë¦¬
            print(f"\n--- {vehicle.name} ìš´í–‰ ---")

            # ì„¼ì„œ ì²´í¬
            if vehicle.check_sensors():  # ifë¬¸ìœ¼ë¡œ ì•ˆì „ì„± í™•ì¸
                vehicle.move_to_destination()
            else:
                print(f"âš ï¸ {vehicle.name}: ì„¼ì„œ ì´ìƒìœ¼ë¡œ ìš´í–‰ ì¤‘ë‹¨")

# í•¨ëŒ€ ë°ì´í„° ì¤€ë¹„ (ë¦¬ìŠ¤íŠ¸ + ë”•ì…”ë„ˆë¦¬ ì¡°í•©)
fleet_data = [
    {
        'name': 'ì„œìš¸íƒì‹œ_001',
        'type': 'sedan',
        'battery': 85,
        'route': [(0, 0), (10, 15)]
    },
    {
        'name': 'ë°°ì†¡íŠ¸ëŸ­_002',
        'type': 'truck',
        'battery': 70,
        'route': [(5, 5), (20, 25)]
    },
    {
        'name': 'ìˆœì°°ì°¨_003',
        'type': 'suv',
        'battery': 95,
        'route': [(2, 3), (8, 12)]
    }
]

# í•¨ëŒ€ ê´€ë¦¬ ì‹œìŠ¤í…œ ìƒì„± ë° ìš´í–‰
fleet_manager = FleetManager("ìŠ¤ë§ˆíŠ¸ëª¨ë¹Œë¦¬í‹°", fleet_data)
fleet_manager.start_fleet_operation()

print("\n" + "="*60)
print("ğŸ“ __init__ ë©”ì„œë“œ í•™ìŠµ ì™„ë£Œ!")
print("ğŸ“š ì§€ê¸ˆê¹Œì§€ ë°°ìš´ ëª¨ë“  ë‚´ìš©ì´ ì¢…í•©ëœ ì˜ˆì œì˜€ìŠµë‹ˆë‹¤:")
print("   âœ… classì™€ __init__ ë©”ì„œë“œ")
print("   âœ… defë¡œ ë©”ì„œë“œ ì •ì˜")
print("   âœ… ifë¬¸ìœ¼ë¡œ ì¡°ê±´ ì²˜ë¦¬")
print("   âœ… forë¬¸ìœ¼ë¡œ ë°˜ë³µ ì²˜ë¦¬")
print("   âœ… ë¦¬ìŠ¤íŠ¸, ë”•ì…”ë„ˆë¦¬, íŠœí”Œ í™œìš©")
print("   âœ… íŠœí”Œ ì–¸íŒ¨í‚¹")
print("   âœ… ì‹¤ì œ ììœ¨ì£¼í–‰ ì‹œë‚˜ë¦¬ì˜¤")
print("="*60)
```
