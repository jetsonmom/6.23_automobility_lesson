```python

# ===== 1Îã®Í≥Ñ: Í∏∞Î≥∏ ÏûêÏú®Ï£ºÌñâÏ∞® ÌÅ¥ÎûòÏä§ =====
print("=== 1Îã®Í≥Ñ: Í∏∞Î≥∏ ÏûêÏú®Ï£ºÌñâÏ∞® ===")

class BasicCar:
    """Í∞ÄÏû• Í∏∞Î≥∏Ï†ÅÏù∏ ÏûêÏú®Ï£ºÌñâÏ∞®"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = 0
        print(f"üöó {self.name} ÏÉùÏÑ±!")

    def move(self, distance):
        if self.battery >= distance:
            self.battery -= distance
            self.position += distance
            print(f"‚úÖ {distance}km Ïù¥Îèô ÏôÑÎ£å! ÏúÑÏπò: {self.position}km, Î∞∞ÌÑ∞Î¶¨: {self.battery}%")
        else:
            print(f"‚ùå Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°±! ÌòÑÏû¨: {self.battery}%, ÌïÑÏöî: {distance}%")

    def charge(self, amount):
        self.battery += amount
        if self.battery > 100:
            self.battery = 100
        print(f"üîã Ï∂©Ï†Ñ ÏôÑÎ£å! Î∞∞ÌÑ∞Î¶¨: {self.battery}%")

# 1Îã®Í≥Ñ ÌÖåÏä§Ìä∏
car1 = BasicCar("ÌÖåÏä¨Îùº")
car1.move(30)
car1.move(50)
car1.charge(40)
car1.move(50)

print("\n" + "="*50 + "\n")
```

```python
# ===== 2Îã®Í≥Ñ: ÏÑºÏÑú Í∏∞Îä• Ï∂îÍ∞Ä =====
print("=== 2Îã®Í≥Ñ: ÏÑºÏÑú Í∏∞Îä• Ï∂îÍ∞Ä ===")

class SensorCar:
    """ÏÑºÏÑúÍ∞Ä ÏûàÎäî ÏûêÏú®Ï£ºÌñâÏ∞®"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = 0
        self.sensors = {
            'camera': True,
            'lidar': True,
            'radar': True
        }
        print(f"üöó {self.name} (ÏÑºÏÑú Ïû•Ï∞©) ÏÉùÏÑ±!")

    def check_sensors(self):
        print(f"üîç {self.name} ÏÑºÏÑú Ï†êÍ≤Ä:")
        working_count = 0

        for sensor_name, is_working in self.sensors.items():
            status = "Ï†ïÏÉÅ" if is_working else "Í≥†Ïû•"
            print(f"  {sensor_name}: {status}")
            if is_working:
                working_count += 1

        if working_count >= 2:
            print("‚úÖ Ï£ºÌñâ Í∞ÄÎä• (ÏÑºÏÑú 2Í∞ú Ïù¥ÏÉÅ Ï†ïÏÉÅ)")
            return True
        else:
            print("‚ùå Ï£ºÌñâ Î∂àÍ∞Ä (ÏÑºÏÑú Î∂ÄÏ°±)")
            return False

    def move_safely(self, distance):
        if self.check_sensors():
            if self.battery >= distance:
                self.battery -= distance
                self.position += distance
                print(f"‚úÖ ÏïàÏ†ÑÌïòÍ≤å {distance}km Ïù¥Îèô! ÏúÑÏπò: {self.position}km")
            else:
                print(f"‚ùå Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°±!")
        else:
            print("‚ö†Ô∏è ÏÑºÏÑú Ïù¥ÏÉÅÏúºÎ°ú Ïù¥Îèô Ï§ëÎã®!")

# 2Îã®Í≥Ñ ÌÖåÏä§Ìä∏
car2 = SensorCar("ÌòÑÎåÄ")
car2.move_safely(20)

# ÏÑºÏÑú Í≥†Ïû• ÏãúÎÆ¨Î†àÏù¥ÏÖò
car2.sensors['camera'] = False
car2.sensors['lidar'] = False
car2.move_safely(20)

print("\n" + "="*50 + "\n")
```

```python
# ===== 3Îã®Í≥Ñ: Ïã†Ìò∏Îì± ÏãúÏä§ÌÖú Ï∂îÍ∞Ä =====
print("=== 3Îã®Í≥Ñ: Ïã†Ìò∏Îì± ÏãúÏä§ÌÖú ===")

class TrafficLight:
    """Ïã†Ìò∏Îì± ÌÅ¥ÎûòÏä§"""

    def __init__(self, name, initial_color='red'):
        self.name = name
        self.color = initial_color
        self.time_left = 30
        print(f"üö¶ {self.name} Ïã†Ìò∏Îì± ÏÑ§Ïπò ({self.color})")

    def get_status(self):
        if self.color == 'green':
            return True  # ÌÜµÍ≥º Í∞ÄÎä•
        else:
            return False  # Ï†ïÏßÄ

    def change_signal(self):
        if self.color == 'red':
            self.color = 'green'
        elif self.color == 'green':
            self.color = 'yellow'
        else:
            self.color = 'red'
        print(f"üö¶ {self.name}: {self.color} Ïã†Ìò∏Î°ú Î≥ÄÍ≤Ω")

class SmartCar:
    """Ïã†Ìò∏Îì±ÏùÑ Ïù∏ÏãùÌïòÎäî ÏûêÏú®Ï£ºÌñâÏ∞®"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = 0
        print(f"üöó {self.name} (Ïä§ÎßàÌä∏) ÏÉùÏÑ±!")

    def approach_intersection(self, traffic_light):
        print(f"\nüöó {self.name}: {traffic_light.name} Ï†ëÍ∑º")
        print(f"üö¶ ÌòÑÏû¨ Ïã†Ìò∏: {traffic_light.color}")

        if traffic_light.get_status():
            print("‚úÖ Ï¥àÎ°ùÎ∂à - ÌÜµÍ≥º!")
            self.position += 10
            self.battery -= 5
        else:
            print("üõë Îπ®Í∞ÑÎ∂à/ÎÖ∏ÎûÄÎ∂à - Ï†ïÏßÄ!")
            print("‚è≥ Ïã†Ìò∏ Î≥ÄÍ≤Ω ÎåÄÍ∏∞ Ï§ë...")

# 3Îã®Í≥Ñ ÌÖåÏä§Ìä∏
light1 = TrafficLight("ÍµêÏ∞®Î°úA", 'red')
car3 = SmartCar("BMW")

car3.approach_intersection(light1)
light1.change_signal()  # Ï¥àÎ°ùÎ∂àÎ°ú Î≥ÄÍ≤Ω
car3.approach_intersection(light1)

print("\n" + "="*50 + "\n")
```

```python
# ===== 4Îã®Í≥Ñ: Í≤ΩÎ°ú Í≥ÑÌöç Í∏∞Îä• =====
print("=== 4Îã®Í≥Ñ: Í≤ΩÎ°ú Í≥ÑÌöç ===")

class NavigationCar:
    """Í≤ΩÎ°ú Í≥ÑÌöçÏù¥ Í∞ÄÎä•Ìïú ÏûêÏú®Ï£ºÌñâÏ∞®"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = [0, 0]  # [x, y] Ï¢åÌëú
        print(f"üöó {self.name} (ÎÇ¥ÎπÑÍ≤åÏù¥ÏÖò) ÏÉùÏÑ±!")

    def plan_route(self, waypoints):
        print(f"\nüó∫Ô∏è {self.name} Í≤ΩÎ°ú Í≥ÑÌöç:")
        total_distance = 0

        current_x, current_y = self.position

        for i, waypoint in enumerate(waypoints):
            target_x, target_y = waypoint
            distance = abs(target_x - current_x) + abs(target_y - current_y)
            total_distance += distance
            print(f"  {i+1}Îã®Í≥Ñ: ({target_x}, {target_y}) - Í±∞Î¶¨ {distance}km")
            current_x, current_y = target_x, target_y

        print(f"üìè Ï¥ù Í±∞Î¶¨: {total_distance}km")
        battery_needed = total_distance * 2

        if self.battery >= battery_needed:
            print("‚úÖ Î∞∞ÌÑ∞Î¶¨ Ï∂©Î∂Ñ - Ï∂úÎ∞ú Í∞ÄÎä•!")
            return True
        else:
            print(f"‚ùå Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°± - ÌïÑÏöî: {battery_needed}%, ÌòÑÏû¨: {self.battery}%")
            return False

    def move_to_waypoint(self, x, y):
        start_x, start_y = self.position
        distance = abs(x - start_x) + abs(y - start_y)
        battery_cost = distance * 2

        if self.battery >= battery_cost:
            self.position = [x, y]
            self.battery -= battery_cost
            print(f"‚úÖ ({x}, {y}) ÎèÑÏ∞©! Î∞∞ÌÑ∞Î¶¨: {self.battery}%")
            return True
        else:
            print(f"‚ùå Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°±ÏúºÎ°ú Ïù¥Îèô Ïã§Ìå®!")
            return False

# 4Îã®Í≥Ñ ÌÖåÏä§Ìä∏
car4 = NavigationCar("Íµ¨Í∏ÄÏπ¥")
route = [(3, 0), (3, 5), (8, 5), (8, 10)]

if car4.plan_route(route):
    print("\nüõ£Ô∏è Í≤ΩÎ°ú Ïã§Ìñâ:")
    for waypoint in route:
        x, y = waypoint
        if not car4.move_to_waypoint(x, y):
            break

print("\n" + "="*50 + "\n")
```

```python
# ClassÏôÄ FunctionÏùÑ ÌôúÏö©Ìïú ÏûêÏú®Ï£ºÌñâÏ∞® ÏãúÎÆ¨Î†àÏù¥ÌÑ∞
# ÏΩîÎû©ÏóêÏÑú Î∞îÎ°ú Ïã§Ìñâ Í∞ÄÎä•!

class AutonomousCar:
    """ÏûêÏú®Ï£ºÌñâÏ∞® ÌÅ¥ÎûòÏä§"""

    def __init__(self, car_id, battery=100):
        """ÏûêÏú®Ï£ºÌñâÏ∞® Ï¥àÍ∏∞Ìôî"""
        self.car_id = car_id
        self.battery = battery
        self.position = [0, 0]  # [x, y] Ï¢åÌëú
        self.speed = 0
        self.destination = None
        self.sensors = {
            'camera': True,
            'lidar': True,
            'radar': True,
            'gps': True
        }
        print(f"üöó {self.car_id} ÏûêÏú®Ï£ºÌñâÏ∞® ÏÉùÏÑ± ÏôÑÎ£å!")

    def check_battery(self):
        """Î∞∞ÌÑ∞Î¶¨ ÏÉÅÌÉú ÌôïÏù∏"""
        if self.battery > 50:
            return "Ï∂©Î∂Ñ"
        elif self.battery > 20:
            return "Î≥¥ÌÜµ"
        else:
            return "Î∂ÄÏ°±"

    def move_to(self, x, y):
        """Î™©Ï†ÅÏßÄÎ°ú Ïù¥Îèô"""
        print(f"\nüéØ {self.car_id}: ({x}, {y})Î°ú Ïù¥Îèô ÏãúÏûë")

        start_x, start_y = self.position
        distance = abs(x - start_x) + abs(y - start_y)  # Îß®ÌïòÌÉÑ Í±∞Î¶¨

        # Î∞∞ÌÑ∞Î¶¨ ÏÜåÎ™® Í≥ÑÏÇ∞
        battery_needed = distance * 2

        if self.battery < battery_needed:
            print(f"‚ùå Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°±! ÌòÑÏû¨: {self.battery}%, ÌïÑÏöî: {battery_needed}%")
            return False

        # Ïù¥Îèô ÏãúÎÆ¨Î†àÏù¥ÏÖò
        self.position = [x, y]
        self.battery -= battery_needed

        print(f"‚úÖ ÎèÑÏ∞©! ÌòÑÏû¨ ÏúÑÏπò: {self.position}")
        print(f"üîã ÎÇ®ÏùÄ Î∞∞ÌÑ∞Î¶¨: {self.battery}%")
        return True

    def scan_sensors(self):
        """ÏÑºÏÑú ÏÉÅÌÉú Ïä§Ï∫î"""
        print(f"\nüîç {self.car_id} ÏÑºÏÑú Ïä§Ï∫î Ï§ë...")

        working_sensors = []
        broken_sensors = []

        for sensor_name, is_working in self.sensors.items():
            if is_working:
                working_sensors.append(sensor_name)
            else:
                broken_sensors.append(sensor_name)

        print(f"‚úÖ Ï†ïÏÉÅ ÏÑºÏÑú: {working_sensors}")
        if broken_sensors:
            print(f"‚ùå Í≥†Ïû• ÏÑºÏÑú: {broken_sensors}")

        return len(working_sensors) >= 3  # 3Í∞ú Ïù¥ÏÉÅÏù¥Î©¥ ÏïàÏ†Ñ

    def plan_route(self, waypoints):
        """Í≤ΩÎ°ú Í≥ÑÌöç"""
        print(f"\nüó∫Ô∏è {self.car_id} Í≤ΩÎ°ú Í≥ÑÌöç:")

        total_distance = 0
        current_pos = self.position.copy()

        for i, waypoint in enumerate(waypoints):
            x, y = waypoint
            distance = abs(x - current_pos[0]) + abs(y - current_pos[1])
            total_distance += distance
            current_pos = [x, y]
            print(f"  {i+1}Îã®Í≥Ñ: {waypoint} (Í±∞Î¶¨: {distance})")

        print(f"üìè Ï¥ù Í±∞Î¶¨: {total_distance}")

        # Î∞∞ÌÑ∞Î¶¨Î°ú Í∞à Ïàò ÏûàÎäîÏßÄ ÌôïÏù∏
        battery_needed = total_distance * 2
        if self.battery >= battery_needed:
            print("‚úÖ Î∞∞ÌÑ∞Î¶¨ Ï∂©Î∂Ñ - Í≤ΩÎ°ú Ïã§Ìñâ Í∞ÄÎä•")
            return True
        else:
            print(f"‚ùå Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°± - ÌïÑÏöî: {battery_needed}%, ÌòÑÏû¨: {self.battery}%")
            return False


class TrafficManager:
    """ÍµêÌÜµ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖú ÌÅ¥ÎûòÏä§"""

    def __init__(self):
        """ÍµêÌÜµ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖú Ï¥àÍ∏∞Ìôî"""
        self.intersections = {
            'ÍµêÏ∞®Î°úA': {'signal': 'green', 'time_left': 30},
            'ÍµêÏ∞®Î°úB': {'signal': 'red', 'time_left': 45},
            'ÍµêÏ∞®Î°úC': {'signal': 'yellow', 'time_left': 5}
        }
        self.cars = []
        print("üö¶ ÍµêÌÜµ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖú ÏãúÏûë!")

    def add_car(self, car):
        """Ï∞®Îüâ Îì±Î°ù"""
        self.cars.append(car)
        print(f"üìù {car.car_id} Ï∞®Îüâ Îì±Î°ù ÏôÑÎ£å")

    def check_traffic_lights(self):
        """Ïã†Ìò∏Îì± ÏÉÅÌÉú ÌôïÏù∏"""
        print("\nüö¶ ÌòÑÏû¨ Ïã†Ìò∏Îì± ÏÉÅÌÉú:")

        for intersection, info in self.intersections.items():
            signal = info['signal']
            time_left = info['time_left']

            if signal == 'green':
                status = "üü¢ ÌÜµÍ≥º Í∞ÄÎä•"
            elif signal == 'yellow':
                status = "üü° Ï£ºÏùò ÌïÑÏöî"
            else:
                status = "üî¥ Ï†ïÏßÄ ÌïÑÏöî"

            print(f"  {intersection}: {status} ({time_left}Ï¥à)")

    def find_safe_route(self, start, end):
        """ÏïàÏ†ÑÌïú Í≤ΩÎ°ú Ï∞æÍ∏∞"""
        print(f"\nüõ£Ô∏è {start} ‚Üí {end} ÏïàÏ†Ñ Í≤ΩÎ°ú Í≤ÄÏÉâ")

        # Ïã†Ìò∏Îì± ÏÉÅÌÉú ÌôïÏù∏Ìï¥ÏÑú Í≤ΩÎ°ú Ï∂îÏ≤ú
        green_lights = []
        red_lights = []

        for intersection, info in self.intersections.items():
            if info['signal'] == 'green':
                green_lights.append(intersection)
            elif info['signal'] == 'red':
                red_lights.append(intersection)

        if green_lights:
            recommended = green_lights[0]
            print(f"‚úÖ Ï∂îÏ≤ú Í≤ΩÎ°ú: {recommended} Í≤ΩÏú† (Ïã†Ìò∏ ÏñëÌò∏)")
        else:
            print("‚ö†Ô∏è Î™®Îì† Ïã†Ìò∏Îì± Ï£ºÏùò - ÏÑúÌñâ Ïö¥Ï†Ñ")

    def monitor_fleet(self):
        """Ï∞®Îüâ Ìï®ÎåÄ Î™®ÎãàÌÑ∞ÎßÅ"""
        print("\nüìä Ï∞®Îüâ Ìï®ÎåÄ ÌòÑÌô©:")

        if not self.cars:
            print("Îì±Î°ùÎêú Ï∞®ÎüâÏù¥ ÏóÜÏäµÎãàÎã§.")
            return

        total_cars = len(self.cars)
        low_battery_cars = []
        safe_cars = []

        for car in self.cars:
            battery_status = car.check_battery()
            sensor_ok = car.scan_sensors()

            print(f"\n{car.car_id}:")
            print(f"  ÏúÑÏπò: {car.position}")
            print(f"  Î∞∞ÌÑ∞Î¶¨: {car.battery}% ({battery_status})")
            print(f"  ÏÑºÏÑú ÏÉÅÌÉú: {'Ï†ïÏÉÅ' if sensor_ok else 'Ï†êÍ≤ÄÌïÑÏöî'}")

            if battery_status == "Î∂ÄÏ°±" or not sensor_ok:
                low_battery_cars.append(car.car_id)
            else:
                safe_cars.append(car.car_id)

        print(f"\nüìà Ìï®ÎåÄ ÏöîÏïΩ:")
        print(f"Ï¥ù Ï∞®Îüâ: {total_cars}ÎåÄ")
        print(f"Ï†ïÏÉÅ Ïö¥Ìñâ: {len(safe_cars)}ÎåÄ")
        print(f"Ï†êÍ≤Ä ÌïÑÏöî: {len(low_battery_cars)}ÎåÄ")

        if low_battery_cars:
            print(f"‚ö†Ô∏è Ï†êÍ≤Ä ÎåÄÏÉÅ: {low_battery_cars}")


def simulate_emergency_scenario():
    """ÏùëÍ∏â ÏÉÅÌô© ÏãúÎÆ¨Î†àÏù¥ÏÖò Ìï®Ïàò"""
    print("\nüö® ÏùëÍ∏â ÏÉÅÌô© Î∞úÏÉù!")

    # ÏùëÍ∏âÏ∞®Îüâ ÏÉùÏÑ±
    ambulance = AutonomousCar("ÏùëÍ∏âÏ∞®_001", battery=80)

    # ÏùëÍ∏â Í≤ΩÎ°ú ÏÑ§Ï†ï
    emergency_route = [(5, 0), (5, 5), (10, 5), (10, 10)]

    print("üè• Î≥ëÏõêÏúºÎ°ú Í∏¥Í∏â Ïù¥ÏÜ° Ï§ë...")

    # Í≤ΩÎ°ú Îî∞Îùº Ïù¥Îèô
    for i, waypoint in enumerate(emergency_route):
        x, y = waypoint
        print(f"\nÎã®Í≥Ñ {i+1}: {waypoint}Î°ú Ïù¥Îèô")

        # Ïù¥Îèô Í∞ÄÎä• Ïó¨Î∂Ä ÌôïÏù∏
        if ambulance.move_to(x, y):
            print(f"‚úÖ Îã®Í≥Ñ {i+1} ÏôÑÎ£å")
        else:
            print(f"‚ùå Îã®Í≥Ñ {i+1} Ïã§Ìå® - ÏùëÍ∏â Ï∂©Ï†Ñ ÌïÑÏöî")
            break

    print("üè• Î≥ëÏõê ÎèÑÏ∞©!")


def run_autonomous_driving_simulation():
    """ÏûêÏú®Ï£ºÌñâ ÏãúÎÆ¨Î†àÏù¥ÏÖò Î©îÏù∏ Ìï®Ïàò"""
    print("=" * 50)
    print("üöó ÏûêÏú®Ï£ºÌñâÏ∞® ÏãúÎÆ¨Î†àÏù¥ÏÖò ÏãúÏûë!")
    print("=" * 50)

    # ÍµêÌÜµ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖú ÏÉùÏÑ±
    traffic_manager = TrafficManager()

    # ÏûêÏú®Ï£ºÌñâÏ∞®Îì§ ÏÉùÏÑ±
    car1 = AutonomousCar("ÏûêÏú®Ï∞®_001", battery=90)
    car2 = AutonomousCar("ÏûêÏú®Ï∞®_002", battery=30)
    car3 = AutonomousCar("ÏûêÏú®Ï∞®_003", battery=75)

    # Ï∞®ÎüâÎì§ÏùÑ ÍµêÌÜµ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖúÏóê Îì±Î°ù
    traffic_manager.add_car(car1)
    traffic_manager.add_car(car2)
    traffic_manager.add_car(car3)

    # Ïã†Ìò∏Îì± ÌôïÏù∏
    traffic_manager.check_traffic_lights()

    # ÏïàÏ†Ñ Í≤ΩÎ°ú Í≤ÄÏÉâ
    traffic_manager.find_safe_route("Ï∂úÎ∞úÏßÄ", "Î™©Ï†ÅÏßÄ")

    # Ï∞®ÎüâÎ≥Ñ Í≤ΩÎ°ú Í≥ÑÌöç Î∞è Ïù¥Îèô
    print("\n" + "=" * 30)
    print("üõ£Ô∏è Í∞úÎ≥Ñ Ï∞®Îüâ Ïö¥Ìñâ ÏãúÏûë")
    print("=" * 30)

    # Ï∞®Îüâ 1: Ï†ïÏÉÅ Ïö¥Ìñâ
    route1 = [(3, 3), (6, 6), (10, 10)]
    if car1.plan_route(route1):
        for waypoint in route1:
            car1.move_to(waypoint[0], waypoint[1])

    # Ï∞®Îüâ 2: Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°±ÏúºÎ°ú Ï†úÌïúÏ†Å Ïù¥Îèô
    route2 = [(2, 2), (4, 4)]
    if car2.plan_route(route2):
        for waypoint in route2:
            if not car2.move_to(waypoint[0], waypoint[1]):
                break

    # Ï∞®Îüâ 3: ÏÑºÏÑú Í≥†Ïû• ÏãúÎÆ¨Î†àÏù¥ÏÖò
    car3.sensors['camera'] = False  # Ïπ¥Î©îÎùº Í≥†Ïû•
    car3.sensors['lidar'] = False   # ÎùºÏù¥Îã§ Í≥†Ïû•

    route3 = [(1, 1), (2, 2)]
    if car3.scan_sensors():  # ÏÑºÏÑú ÌôïÏù∏
        car3.plan_route(route3)
    else:
        print(f"‚ö†Ô∏è {car3.car_id}: ÏÑºÏÑú Ïù¥ÏÉÅÏúºÎ°ú Ïö¥Ìñâ Ï§ëÎã®")

    # Ï†ÑÏ≤¥ Ìï®ÎåÄ Î™®ÎãàÌÑ∞ÎßÅ
    traffic_manager.monitor_fleet()

    # ÏùëÍ∏â ÏÉÅÌô© ÏãúÎÆ¨Î†àÏù¥ÏÖò
    simulate_emergency_scenario()

    print("\n" + "=" * 50)
    print("üèÅ ÏãúÎÆ¨Î†àÏù¥ÏÖò ÏôÑÎ£å!")
    print("=" * 50)


# ÏãúÎÆ¨Î†àÏù¥ÏÖò Ïã§Ìñâ
if __name__ == "__main__":
    run_autonomous_driving_simulation()
```

```python
# __init__ Î©îÏÑúÎìú ÏôÑÏ†Ñ Í∞ÄÏù¥Îìú with ÏûêÏú®Ï£ºÌñâ Ï¢ÖÌï© ÏòàÏ†ú

# ===== __init__ Í∏∞Î≥∏ Ïù¥Ìï¥ =====
print("=== __init__ Î©îÏÑúÎìúÎûÄ? ===")
print("üîß __init__ÏùÄ 'ÏÉùÏÑ±Ïûê'ÎùºÍ≥† Î∂àÎ¶¨Îäî ÌäπÎ≥ÑÌïú Î©îÏÑúÎìúÏûÖÎãàÎã§")
print("üéØ ÌÅ¥ÎûòÏä§Î°úÎ∂ÄÌÑ∞ Í∞ùÏ≤¥Î•º ÏÉùÏÑ±Ìï† Îïå 'ÏûêÎèôÏúºÎ°ú' Ïã§ÌñâÎê©ÎãàÎã§")
print("üìù Í∞ùÏ≤¥Ïùò 'Ï¥àÍ∏∞ ÏÉÅÌÉú'Î•º ÏÑ§Ï†ïÌïòÎäî Ïó≠Ìï†ÏùÑ Ìï©ÎãàÎã§")
print()

# ===== 1Îã®Í≥Ñ: Í∞ÄÏû• Í∏∞Î≥∏Ï†ÅÏù∏ __init__ =====
print("=== 1Îã®Í≥Ñ: Í∏∞Î≥∏ __init__ ===")

class SimpleCar:
    def __init__(self):  # Îß§Í∞úÎ≥ÄÏàò ÏóÜÎäî Í∏∞Î≥∏Ìòï
        print("üöó SimpleCar __init__ Ïã§ÌñâÎê®!")
        self.speed = 0
        self.engine = "off"
        print(f"   Ï¥àÍ∏∞ ÏÜçÎèÑ: {self.speed}")
        print(f"   Ï¥àÍ∏∞ ÏóîÏßÑ ÏÉÅÌÉú: {self.engine}")

# Í∞ùÏ≤¥ ÏÉùÏÑ± - __init__Ïù¥ ÏûêÎèô Ïã§ÌñâÎê®
print("car1 = SimpleCar() Ïã§Ìñâ:")
car1 = SimpleCar()
print()

# ===== 2Îã®Í≥Ñ: Îß§Í∞úÎ≥ÄÏàòÍ∞Ä ÏûàÎäî __init__ =====
print("=== 2Îã®Í≥Ñ: Îß§Í∞úÎ≥ÄÏàò Î∞õÎäî __init__ ===")

class BasicAutoCar:
    def __init__(self, car_name, initial_battery):  # Îß§Í∞úÎ≥ÄÏàò Î∞õÍ∏∞
        print(f"üöó {car_name} ÏûêÎèôÏ∞® ÏÉùÏÑ± Ï§ë...")

        # self.ÏÜçÏÑ±Î™Ö = Í∞í ÏúºÎ°ú Í∞ùÏ≤¥Ïùò ÏÜçÏÑ± ÏÑ§Ï†ï
        self.name = car_name
        self.battery = initial_battery
        self.position = [0, 0]  # Î¶¨Ïä§Ìä∏ ÏÇ¨Ïö©
        self.sensors = {        # ÎîïÏÖîÎÑàÎ¶¨ ÏÇ¨Ïö©
            'camera': True,
            'lidar': True,
            'radar': False
        }

        print(f"   Ïù¥Î¶Ñ: {self.name}")
        print(f"   Î∞∞ÌÑ∞Î¶¨: {self.battery}%")
        print(f"   ÏúÑÏπò: {self.position}")
        print(f"   ÏÑºÏÑú ÏÉÅÌÉú: {self.sensors}")
        print("‚úÖ Ï¥àÍ∏∞Ìôî ÏôÑÎ£å!")

# Îß§Í∞úÎ≥ÄÏàòÎ•º Ï£ºÎ©¥ÏÑú Í∞ùÏ≤¥ ÏÉùÏÑ±
print("car2 = BasicAutoCar('ÌÖåÏä¨Îùº', 85) Ïã§Ìñâ:")
car2 = BasicAutoCar('ÌÖåÏä¨Îùº', 85)
print()

print("car3 = BasicAutoCar('ÌòÑÎåÄ', 60) Ïã§Ìñâ:")
car3 = BasicAutoCar('ÌòÑÎåÄ', 60)
print()

# ===== 3Îã®Í≥Ñ: Î≥µÌï© Îç∞Ïù¥ÌÑ∞Î•º Î∞õÎäî __init__ =====
print("=== 3Îã®Í≥Ñ: Î≥µÌï© Îç∞Ïù¥ÌÑ∞ Î∞õÎäî __init__ ===")

class AdvancedAutoCar:
    def __init__(self, car_info, sensor_list, route_data):
        """
        car_info: ÌäúÌîå (Ïù¥Î¶Ñ, Î∞∞ÌÑ∞Î¶¨, ÌÉÄÏûÖ)
        sensor_list: Î¶¨Ïä§Ìä∏ ['camera', 'lidar', 'radar']
        route_data: ÎîïÏÖîÎÑàÎ¶¨ {'start': (0,0), 'end': (10,10)}
        """
        print("üöó Í≥†Í∏â ÏûêÏú®Ï£ºÌñâÏ∞® ÏÉùÏÑ± Ï§ë...")

        # ÌäúÌîå Ïñ∏Ìå®ÌÇπÏúºÎ°ú Ï†ïÎ≥¥ Î∂ÑÎ¶¨
        name, battery, car_type = car_info
        self.name = name
        self.battery = battery
        self.car_type = car_type

        # Î¶¨Ïä§Ìä∏Î•º ÎîïÏÖîÎÑàÎ¶¨Î°ú Î≥ÄÌôò (Î™®Îì† ÏÑºÏÑúÎ•º TrueÎ°ú ÏÑ§Ï†ï)
        self.sensors = {}
        for sensor in sensor_list:  # forÎ¨∏ ÏÇ¨Ïö©
            self.sensors[sensor] = True

        # ÎîïÏÖîÎÑàÎ¶¨ Ï†ïÎ≥¥ Ï†ÄÏû•
        self.start_position = route_data['start']
        self.end_position = route_data['end']
        self.current_position = list(self.start_position)  # Î¶¨Ïä§Ìä∏Î°ú Î≥µÏÇ¨

        # Ï∂îÍ∞Ä Í≥ÑÏÇ∞
        start_x, start_y = self.start_position  # ÌäúÌîå Ïñ∏Ìå®ÌÇπ
        end_x, end_y = self.end_position
        self.total_distance = abs(end_x - start_x) + abs(end_y - start_y)

        print(f"   Ï∞®ÎüâÎ™Ö: {self.name} ({self.car_type})")
        print(f"   Î∞∞ÌÑ∞Î¶¨: {self.battery}%")
        print(f"   ÏÑºÏÑú: {list(self.sensors.keys())}")
        print(f"   Ï∂úÎ∞úÏßÄ: {self.start_position}")
        print(f"   Î™©Ï†ÅÏßÄ: {self.end_position}")
        print(f"   Ï¥ù Í±∞Î¶¨: {self.total_distance}km")
        print("‚úÖ Í≥†Í∏â Ï¥àÍ∏∞Ìôî ÏôÑÎ£å!")

    def move_to_destination(self):
        """Î™©Ï†ÅÏßÄÎ°ú Ïù¥ÎèôÌïòÎäî Î©îÏÑúÎìú"""
        print(f"\nüõ£Ô∏è {self.name} Î™©Ï†ÅÏßÄÎ°ú Ïù¥Îèô ÏãúÏûë!")

        # ifÎ¨∏ÏúºÎ°ú Î∞∞ÌÑ∞Î¶¨ ÌôïÏù∏
        battery_needed = self.total_distance * 2
        if self.battery >= battery_needed:
            self.current_position = list(self.end_position)
            self.battery -= battery_needed
            print(f"‚úÖ Î™©Ï†ÅÏßÄ ÎèÑÏ∞©! ÌòÑÏû¨ ÏúÑÏπò: {self.current_position}")
            print(f"üîã ÎÇ®ÏùÄ Î∞∞ÌÑ∞Î¶¨: {self.battery}%")
        else:
            print(f"‚ùå Î∞∞ÌÑ∞Î¶¨ Î∂ÄÏ°±! ÌïÑÏöî: {battery_needed}%, ÌòÑÏû¨: {self.battery}%")

    def check_sensors(self):
        """ÏÑºÏÑú ÏÉÅÌÉú Ï≤¥ÌÅ¨"""
        print(f"\nüîç {self.name} ÏÑºÏÑú Ï†êÍ≤Ä:")
        working_sensors = []

        for sensor_name, is_working in self.sensors.items():  # ÎîïÏÖîÎÑàÎ¶¨ ÏàúÌöå
            if is_working:  # ifÎ¨∏ÏúºÎ°ú ÏÉÅÌÉú ÌôïÏù∏
                working_sensors.append(sensor_name)
                print(f"  ‚úÖ {sensor_name}: Ï†ïÏÉÅ")
            else:
                print(f"  ‚ùå {sensor_name}: Í≥†Ïû•")

        return len(working_sensors) >= 2  # 2Í∞ú Ïù¥ÏÉÅÏù¥Î©¥ ÏïàÏ†Ñ

# Î≥µÌï© Îç∞Ïù¥ÌÑ∞Î°ú Í∞ùÏ≤¥ ÏÉùÏÑ±
car_info = ('BMW X5', 90, 'SUV')  # ÌäúÌîå
sensor_list = ['camera', 'lidar', 'radar', 'ultrasonic']  # Î¶¨Ïä§Ìä∏
route_data = {'start': (0, 0), 'end': (15, 20)}  # ÎîïÏÖîÎÑàÎ¶¨

print("Í≥†Í∏â ÏûêÏú®Ï£ºÌñâÏ∞® ÏÉùÏÑ±:")
advanced_car = AdvancedAutoCar(car_info, sensor_list, route_data)

# Î©îÏÑúÎìú ÏÇ¨Ïö©
advanced_car.check_sensors()
advanced_car.move_to_destination()
print()

# ===== 4Îã®Í≥Ñ: Ï°∞Í±¥Î∂Ä Ï¥àÍ∏∞ÌôîÍ∞Ä ÏûàÎäî __init__ =====
print("=== 4Îã®Í≥Ñ: Ï°∞Í±¥Î∂Ä Ï¥àÍ∏∞Ìôî ===")

class SmartAutoCar:
    def __init__(self, name, car_class='economy', weather='clear'):
        """
        Ï∞®Îüâ Îì±Í∏âÍ≥º ÎÇ†Ïî®Ïóê Îî∞Îùº Îã§Î•¥Í≤å Ï¥àÍ∏∞Ìôî
        """
        print(f"üöó {name} Ïä§ÎßàÌä∏Ïπ¥ ÏÉùÏÑ± Ï§ë... (Îì±Í∏â: {car_class}, ÎÇ†Ïî®: {weather})")

        self.name = name
        self.weather = weather

        # ifÎ¨∏ÏúºÎ°ú Ï∞®Îüâ Îì±Í∏âÏóê Îî∞Î•∏ Îã§Î•∏ Ï¥àÍ∏∞Ìôî
        if car_class == 'economy':
            self.battery = 60
            self.max_speed = 80
            self.sensors = ['camera', 'basic_radar']
        elif car_class == 'premium':
            self.battery = 90
            self.max_speed = 120
            self.sensors = ['camera', 'lidar', 'radar', 'ultrasonic']
        elif car_class == 'luxury':
            self.battery = 100
            self.max_speed = 150
            self.sensors = ['camera', 'lidar', 'radar', 'ultrasonic', 'thermal']
        else:
            # Í∏∞Î≥∏Í∞í ÏÑ§Ï†ï
            self.battery = 50
            self.max_speed = 60
            self.sensors = ['camera']

        # ÎÇ†Ïî®Ïóê Îî∞Î•∏ Ï∂îÍ∞Ä ÏÑ§Ï†ï
        if weather == 'rain':
            self.max_speed = int(self.max_speed * 0.7)  # 30% Í∞êÏÜç
            self.safety_distance = 2.0
        elif weather == 'snow':
            self.max_speed = int(self.max_speed * 0.5)  # 50% Í∞êÏÜç
            self.safety_distance = 3.0
        else:  # clear weather
            self.safety_distance = 1.0

        # ÏÑºÏÑúÎ•º ÎîïÏÖîÎÑàÎ¶¨Î°ú Î≥ÄÌôò
        self.sensor_status = {}
        for sensor in self.sensors:  # forÎ¨∏ÏúºÎ°ú ÏÑºÏÑú ÎîïÏÖîÎÑàÎ¶¨ ÏÉùÏÑ±
            self.sensor_status[sensor] = True

        print(f"   Î∞∞ÌÑ∞Î¶¨: {self.battery}%")
        print(f"   ÏµúÎåÄÏÜçÎèÑ: {self.max_speed}km/h")
        print(f"   ÏïàÏ†ÑÍ±∞Î¶¨: {self.safety_distance}Î∞∞")
        print(f"   ÏÑºÏÑú: {self.sensors}")
        print("‚úÖ Ïä§ÎßàÌä∏ Ï¥àÍ∏∞Ìôî ÏôÑÎ£å!")

    def get_driving_recommendation(self):
        """Ï£ºÌñâ Í∂åÏû•ÏÇ¨Ìï≠ Ï†úÍ≥µ"""
        print(f"\nüìã {self.name} Ï£ºÌñâ Í∂åÏû•ÏÇ¨Ìï≠:")
        print(f"ÌòÑÏû¨ ÎÇ†Ïî®: {self.weather}")
        print(f"Í∂åÏû• ÏµúÎåÄÏÜçÎèÑ: {self.max_speed}km/h")
        print(f"ÏïàÏ†ÑÍ±∞Î¶¨ Î∞∞Ïàò: {self.safety_distance}")

        # ÏÑºÏÑú ÏÉÅÌÉú ÌôïÏù∏
        active_sensors = len([s for s in self.sensor_status.values() if s])
        print(f"ÌôúÏÑ± ÏÑºÏÑú: {active_sensors}Í∞ú")

        # ifÎ¨∏ÏúºÎ°ú Ï¢ÖÌï© Í∂åÏû•ÏÇ¨Ìï≠
        if self.weather == 'clear' and active_sensors >= 3:
            print("‚úÖ Ï†ïÏÉÅ Ïö¥Ìñâ Í∞ÄÎä•")
        elif self.weather in ['rain', 'snow']:
            print("‚ö†Ô∏è ÏÑúÌñâ Ïö¥Ï†Ñ ÌïÑÏàò")
        else:
            print("üî∂ Ï£ºÏùò Ïö¥Ï†Ñ ÌïÑÏöî")

# Îã§ÏñëÌïú Ï°∞Í±¥ÏúºÎ°ú Í∞ùÏ≤¥ ÏÉùÏÑ±
print("economy Îì±Í∏â, ÎßëÏùÄ ÎÇ†Ïî®:")
car_eco = SmartAutoCar("Í≤ΩÏ†úÌòïÏπ¥", 'economy', 'clear')
car_eco.get_driving_recommendation()
print()

print("luxury Îì±Í∏â, ÎπÑÏò§Îäî ÎÇ†Ïî®:")
car_luxury = SmartAutoCar("Îü≠ÏÖîÎ¶¨Ïπ¥", 'luxury', 'rain')
car_luxury.get_driving_recommendation()
print()

# ===== 5Îã®Í≥Ñ: Ï¢ÖÌï© ÏùëÏö© - ÏûêÏú®Ï£ºÌñâ Ìï®ÎåÄ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖú =====
print("=== 5Îã®Í≥Ñ: Ï¢ÖÌï© ÏùëÏö© - Ìï®ÎåÄ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖú ===")

class FleetManager:
    """ÏûêÏú®Ï£ºÌñâ Ìï®ÎåÄ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖú"""

    def __init__(self, company_name, fleet_data):
        """
        fleet_data: Î¶¨Ïä§Ìä∏ [
            {'name': 'Car1', 'type': 'sedan', 'battery': 80, 'route': [(0,0), (10,5)]},
            {'name': 'Car2', 'type': 'suv', 'battery': 60, 'route': [(5,5), (15,10)]}
        ]
        """
        print(f"üè¢ {company_name} Ìï®ÎåÄ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖú Ï¥àÍ∏∞Ìôî Ï§ë...")

        self.company = company_name
        self.vehicles = []  # Ï∞®Îüâ Í∞ùÏ≤¥Îì§ÏùÑ Ï†ÄÏû•Ìï† Î¶¨Ïä§Ìä∏
        self.total_vehicles = len(fleet_data)

        # forÎ¨∏ÏúºÎ°ú Í∞Å Ï∞®Îüâ Îç∞Ïù¥ÌÑ∞Î•º Ï≤òÎ¶¨
        for i, vehicle_data in enumerate(fleet_data):
            print(f"\nÏ∞®Îüâ {i+1}/{self.total_vehicles} Îì±Î°ù Ï§ë:")

            # ÎîïÏÖîÎÑàÎ¶¨ÏóêÏÑú Îç∞Ïù¥ÌÑ∞ Ï∂îÏ∂ú
            name = vehicle_data['name']
            car_type = vehicle_data['type']
            battery = vehicle_data['battery']
            route = vehicle_data['route']

            # Í≤ΩÎ°ú Ï†ïÎ≥¥ Ï≤òÎ¶¨ (ÌäúÌîå Ïñ∏Ìå®ÌÇπ)
            start_point, end_point = route[0], route[-1]

            # Ï∞®Îüâ Ï†ïÎ≥¥Î•º ÌäúÌîåÎ°ú Íµ¨ÏÑ±
            car_info = (name, battery, car_type)

            # Í≤ΩÎ°ú Ï†ïÎ≥¥Î•º ÎîïÏÖîÎÑàÎ¶¨Î°ú Íµ¨ÏÑ±
            route_info = {'start': start_point, 'end': end_point}

            # Ï∞®Îüâ ÌÉÄÏûÖÏóê Îî∞Î•∏ ÏÑºÏÑú ÏÑ§Ï†ï
            if car_type == 'sedan':
                sensors = ['camera', 'radar']
            elif car_type == 'suv':
                sensors = ['camera', 'lidar', 'radar']
            else:  # truck Îì±
                sensors = ['camera', 'lidar', 'radar', 'ultrasonic']

            # AdvancedAutoCar Í∞ùÏ≤¥ ÏÉùÏÑ±
            vehicle = AdvancedAutoCar(car_info, sensors, route_info)
            self.vehicles.append(vehicle)  # Î¶¨Ïä§Ìä∏Ïóê Ï∂îÍ∞Ä

        print(f"\n‚úÖ {self.company} Ìï®ÎåÄ Ï¥àÍ∏∞Ìôî ÏôÑÎ£å!")
        print(f"Ï¥ù Îì±Î°ù Ï∞®Îüâ: {self.total_vehicles}ÎåÄ")

    def start_fleet_operation(self):
        """Ìï®ÎåÄ Ïö¥Ìñâ ÏãúÏûë"""
        print(f"\nüöÄ {self.company} Ìï®ÎåÄ Ïö¥Ìñâ ÏãúÏûë!")

        for vehicle in self.vehicles:  # forÎ¨∏ÏúºÎ°ú Î™®Îì† Ï∞®Îüâ Ï≤òÎ¶¨
            print(f"\n--- {vehicle.name} Ïö¥Ìñâ ---")

            # ÏÑºÏÑú Ï≤¥ÌÅ¨
            if vehicle.check_sensors():  # ifÎ¨∏ÏúºÎ°ú ÏïàÏ†ÑÏÑ± ÌôïÏù∏
                vehicle.move_to_destination()
            else:
                print(f"‚ö†Ô∏è {vehicle.name}: ÏÑºÏÑú Ïù¥ÏÉÅÏúºÎ°ú Ïö¥Ìñâ Ï§ëÎã®")

# Ìï®ÎåÄ Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ (Î¶¨Ïä§Ìä∏ + ÎîïÏÖîÎÑàÎ¶¨ Ï°∞Ìï©)
fleet_data = [
    {
        'name': 'ÏÑúÏö∏ÌÉùÏãú_001',
        'type': 'sedan',
        'battery': 85,
        'route': [(0, 0), (10, 15)]
    },
    {
        'name': 'Î∞∞ÏÜ°Ìä∏Îü≠_002',
        'type': 'truck',
        'battery': 70,
        'route': [(5, 5), (20, 25)]
    },
    {
        'name': 'ÏàúÏ∞∞Ï∞®_003',
        'type': 'suv',
        'battery': 95,
        'route': [(2, 3), (8, 12)]
    }
]

# Ìï®ÎåÄ Í¥ÄÎ¶¨ ÏãúÏä§ÌÖú ÏÉùÏÑ± Î∞è Ïö¥Ìñâ
fleet_manager = FleetManager("Ïä§ÎßàÌä∏Î™®ÎπåÎ¶¨Ìã∞", fleet_data)
fleet_manager.start_fleet_operation()

print("\n" + "="*60)
print("üéì __init__ Î©îÏÑúÎìú ÌïôÏäµ ÏôÑÎ£å!")
print("üìö ÏßÄÍ∏àÍπåÏßÄ Î∞∞Ïö¥ Î™®Îì† ÎÇ¥Ïö©Ïù¥ Ï¢ÖÌï©Îêú ÏòàÏ†úÏòÄÏäµÎãàÎã§:")
print("   ‚úÖ classÏôÄ __init__ Î©îÏÑúÎìú")
print("   ‚úÖ defÎ°ú Î©îÏÑúÎìú Ï†ïÏùò")
print("   ‚úÖ ifÎ¨∏ÏúºÎ°ú Ï°∞Í±¥ Ï≤òÎ¶¨")
print("   ‚úÖ forÎ¨∏ÏúºÎ°ú Î∞òÎ≥µ Ï≤òÎ¶¨")
print("   ‚úÖ Î¶¨Ïä§Ìä∏, ÎîïÏÖîÎÑàÎ¶¨, ÌäúÌîå ÌôúÏö©")
print("   ‚úÖ ÌäúÌîå Ïñ∏Ìå®ÌÇπ")
print("   ‚úÖ Ïã§Ï†ú ÏûêÏú®Ï£ºÌñâ ÏãúÎÇòÎ¶¨Ïò§")
print("="*60)
```
