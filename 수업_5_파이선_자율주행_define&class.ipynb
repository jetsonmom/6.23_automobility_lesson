```python

# ===== 1ë‹¨ê³„: ê¸°ë³¸ ììœ¨ì£¼í–‰ì°¨ í´ë˜ìŠ¤ =====
print("=== 1ë‹¨ê³„: ê¸°ë³¸ ììœ¨ì£¼í–‰ì°¨ ===")

class BasicCar:
    """ê°€ì¥ ê¸°ë³¸ì ì¸ ììœ¨ì£¼í–‰ì°¨"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = 0
        print(f"ğŸš— {self.name} ìƒì„±!")

    def move(self, distance):
        if self.battery >= distance:
            self.battery -= distance
            self.position += distance
            print(f"âœ… {distance}km ì´ë™ ì™„ë£Œ! ìœ„ì¹˜: {self.position}km, ë°°í„°ë¦¬: {self.battery}%")
        else:
            print(f"âŒ ë°°í„°ë¦¬ ë¶€ì¡±! í˜„ì¬: {self.battery}%, í•„ìš”: {distance}%")

    def charge(self, amount):
        self.battery += amount
        if self.battery > 100:
            self.battery = 100
        print(f"ğŸ”‹ ì¶©ì „ ì™„ë£Œ! ë°°í„°ë¦¬: {self.battery}%")

# 1ë‹¨ê³„ í…ŒìŠ¤íŠ¸
car1 = BasicCar("í…ŒìŠ¬ë¼")
car1.move(30)
car1.move(50)
car1.charge(40)
car1.move(50)

print("\n" + "="*50 + "\n")
```

```python
# ===== 2ë‹¨ê³„: ì„¼ì„œ ê¸°ëŠ¥ ì¶”ê°€ =====
print("=== 2ë‹¨ê³„: ì„¼ì„œ ê¸°ëŠ¥ ì¶”ê°€ ===")

class SensorCar:
    """ì„¼ì„œê°€ ìˆëŠ” ììœ¨ì£¼í–‰ì°¨"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = 0
        self.sensors = {
            'camera': True,
            'lidar': True,
            'radar': True
        }
        print(f"ğŸš— {self.name} (ì„¼ì„œ ì¥ì°©) ìƒì„±!")

    def check_sensors(self):
        print(f"ğŸ” {self.name} ì„¼ì„œ ì ê²€:")
        working_count = 0

        for sensor_name, is_working in self.sensors.items():
            status = "ì •ìƒ" if is_working else "ê³ ì¥"
            print(f"  {sensor_name}: {status}")
            if is_working:
                working_count += 1

        if working_count >= 2:
            print("âœ… ì£¼í–‰ ê°€ëŠ¥ (ì„¼ì„œ 2ê°œ ì´ìƒ ì •ìƒ)")
            return True
        else:
            print("âŒ ì£¼í–‰ ë¶ˆê°€ (ì„¼ì„œ ë¶€ì¡±)")
            return False

    def move_safely(self, distance):
        if self.check_sensors():
            if self.battery >= distance:
                self.battery -= distance
                self.position += distance
                print(f"âœ… ì•ˆì „í•˜ê²Œ {distance}km ì´ë™! ìœ„ì¹˜: {self.position}km")
            else:
                print(f"âŒ ë°°í„°ë¦¬ ë¶€ì¡±!")
        else:
            print("âš ï¸ ì„¼ì„œ ì´ìƒìœ¼ë¡œ ì´ë™ ì¤‘ë‹¨!")

# 2ë‹¨ê³„ í…ŒìŠ¤íŠ¸
car2 = SensorCar("í˜„ëŒ€")
car2.move_safely(20)

# ì„¼ì„œ ê³ ì¥ ì‹œë®¬ë ˆì´ì…˜
car2.sensors['camera'] = False
car2.sensors['lidar'] = False
car2.move_safely(20)

print("\n" + "="*50 + "\n")
```

```python
# ===== 3ë‹¨ê³„: ì‹ í˜¸ë“± ì‹œìŠ¤í…œ ì¶”ê°€ =====
print("=== 3ë‹¨ê³„: ì‹ í˜¸ë“± ì‹œìŠ¤í…œ ===")

class TrafficLight:
    """ì‹ í˜¸ë“± í´ë˜ìŠ¤"""

    def __init__(self, name, initial_color='red'):
        self.name = name
        self.color = initial_color
        self.time_left = 30
        print(f"ğŸš¦ {self.name} ì‹ í˜¸ë“± ì„¤ì¹˜ ({self.color})")

    def get_status(self):
        if self.color == 'green':
            return True  # í†µê³¼ ê°€ëŠ¥
        else:
            return False  # ì •ì§€

    def change_signal(self):
        if self.color == 'red':
            self.color = 'green'
        elif self.color == 'green':
            self.color = 'yellow'
        else:
            self.color = 'red'
        print(f"ğŸš¦ {self.name}: {self.color} ì‹ í˜¸ë¡œ ë³€ê²½")

class SmartCar:
    """ì‹ í˜¸ë“±ì„ ì¸ì‹í•˜ëŠ” ììœ¨ì£¼í–‰ì°¨"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = 0
        print(f"ğŸš— {self.name} (ìŠ¤ë§ˆíŠ¸) ìƒì„±!")

    def approach_intersection(self, traffic_light):
        print(f"\nğŸš— {self.name}: {traffic_light.name} ì ‘ê·¼")
        print(f"ğŸš¦ í˜„ì¬ ì‹ í˜¸: {traffic_light.color}")

        if traffic_light.get_status():
            print("âœ… ì´ˆë¡ë¶ˆ - í†µê³¼!")
            self.position += 10
            self.battery -= 5
        else:
            print("ğŸ›‘ ë¹¨ê°„ë¶ˆ/ë…¸ë€ë¶ˆ - ì •ì§€!")
            print("â³ ì‹ í˜¸ ë³€ê²½ ëŒ€ê¸° ì¤‘...")

# 3ë‹¨ê³„ í…ŒìŠ¤íŠ¸
light1 = TrafficLight("êµì°¨ë¡œA", 'red')
car3 = SmartCar("BMW")

car3.approach_intersection(light1)
light1.change_signal()  # ì´ˆë¡ë¶ˆë¡œ ë³€ê²½
car3.approach_intersection(light1)

print("\n" + "="*50 + "\n")
```

```python
# ===== 4ë‹¨ê³„: ê²½ë¡œ ê³„íš ê¸°ëŠ¥ =====
print("=== 4ë‹¨ê³„: ê²½ë¡œ ê³„íš ===")

class NavigationCar:
    """ê²½ë¡œ ê³„íšì´ ê°€ëŠ¥í•œ ììœ¨ì£¼í–‰ì°¨"""

    def __init__(self, name):
        self.name = name
        self.battery = 100
        self.position = [0, 0]  # [x, y] ì¢Œí‘œ
        print(f"ğŸš— {self.name} (ë‚´ë¹„ê²Œì´ì…˜) ìƒì„±!")

    def plan_route(self, waypoints):
        print(f"\nğŸ—ºï¸ {self.name} ê²½ë¡œ ê³„íš:")
        total_distance = 0

        current_x, current_y = self.position

        for i, waypoint in enumerate(waypoints):
            target_x, target_y = waypoint
            distance = abs(target_x - current_x) + abs(target_y - current_y)
            total_distance += distance
            print(f"  {i+1}ë‹¨ê³„: ({target_x}, {target_y}) - ê±°ë¦¬ {distance}km")
            current_x, current_y = target_x, target_y

        print(f"ğŸ“ ì´ ê±°ë¦¬: {total_distance}km")
        battery_needed = total_distance * 2

        if self.battery >= battery_needed:
            print("âœ… ë°°í„°ë¦¬ ì¶©ë¶„ - ì¶œë°œ ê°€ëŠ¥!")
            return True
        else:
            print(f"âŒ ë°°í„°ë¦¬ ë¶€ì¡± - í•„ìš”: {battery_needed}%, í˜„ì¬: {self.battery}%")
            return False

    def move_to_waypoint(self, x, y):
        start_x, start_y = self.position
        distance = abs(x - start_x) + abs(y - start_y)
        battery_cost = distance * 2

        if self.battery >= battery_cost:
            self.position = [x, y]
            self.battery -= battery_cost
            print(f"âœ… ({x}, {y}) ë„ì°©! ë°°í„°ë¦¬: {self.battery}%")
            return True
        else:
            print(f"âŒ ë°°í„°ë¦¬ ë¶€ì¡±ìœ¼ë¡œ ì´ë™ ì‹¤íŒ¨!")
            return False

# 4ë‹¨ê³„ í…ŒìŠ¤íŠ¸
car4 = NavigationCar("êµ¬ê¸€ì¹´")
route = [(3, 0), (3, 5), (8, 5), (8, 10)]

if car4.plan_route(route):
    print("\nğŸ›£ï¸ ê²½ë¡œ ì‹¤í–‰:")
    for waypoint in route:
        x, y = waypoint
        if not car4.move_to_waypoint(x, y):
            break

print("\n" + "="*50 + "\n")
```

```python
# Classì™€ Functionì„ í™œìš©í•œ ììœ¨ì£¼í–‰ì°¨ ì‹œë®¬ë ˆì´í„°
# ì½”ë©ì—ì„œ ë°”ë¡œ ì‹¤í–‰ ê°€ëŠ¥!

class AutonomousCar:
    """ììœ¨ì£¼í–‰ì°¨ í´ë˜ìŠ¤"""

    def __init__(self, car_id, battery=100):
        """ììœ¨ì£¼í–‰ì°¨ ì´ˆê¸°í™”"""
        self.car_id = car_id
        self.battery = battery
        self.position = [0, 0]  # [x, y] ì¢Œí‘œ
        self.speed = 0
        self.destination = None
        self.sensors = {
            'camera': True,
            'lidar': True,
            'radar': True,
            'gps': True
        }
        print(f"ğŸš— {self.car_id} ììœ¨ì£¼í–‰ì°¨ ìƒì„± ì™„ë£Œ!")

    def check_battery(self):
        """ë°°í„°ë¦¬ ìƒíƒœ í™•ì¸"""
        if self.battery > 50:
            return "ì¶©ë¶„"
        elif self.battery > 20:
            return "ë³´í†µ"
        else:
            return "ë¶€ì¡±"

    def move_to(self, x, y):
        """ëª©ì ì§€ë¡œ ì´ë™"""
        print(f"\nğŸ¯ {self.car_id}: ({x}, {y})ë¡œ ì´ë™ ì‹œì‘")

        start_x, start_y = self.position
        distance = abs(x - start_x) + abs(y - start_y)  # ë§¨í•˜íƒ„ ê±°ë¦¬

        # ë°°í„°ë¦¬ ì†Œëª¨ ê³„ì‚°
        battery_needed = distance * 2

        if self.battery < battery_needed:
            print(f"âŒ ë°°í„°ë¦¬ ë¶€ì¡±! í˜„ì¬: {self.battery}%, í•„ìš”: {battery_needed}%")
            return False

        # ì´ë™ ì‹œë®¬ë ˆì´ì…˜
        self.position = [x, y]
        self.battery -= battery_needed

        print(f"âœ… ë„ì°©! í˜„ì¬ ìœ„ì¹˜: {self.position}")
        print(f"ğŸ”‹ ë‚¨ì€ ë°°í„°ë¦¬: {self.battery}%")
        return True

    def scan_sensors(self):
        """ì„¼ì„œ ìƒíƒœ ìŠ¤ìº”"""
        print(f"\nğŸ” {self.car_id} ì„¼ì„œ ìŠ¤ìº” ì¤‘...")

        working_sensors = []
        broken_sensors = []

        for sensor_name, is_working in self.sensors.items():
            if is_working:
                working_sensors.append(sensor_name)
            else:
                broken_sensors.append(sensor_name)

        print(f"âœ… ì •ìƒ ì„¼ì„œ: {working_sensors}")
        if broken_sensors:
            print(f"âŒ ê³ ì¥ ì„¼ì„œ: {broken_sensors}")

        return len(working_sensors) >= 3  # 3ê°œ ì´ìƒì´ë©´ ì•ˆì „

    def plan_route(self, waypoints):
        """ê²½ë¡œ ê³„íš"""
        print(f"\nğŸ—ºï¸ {self.car_id} ê²½ë¡œ ê³„íš:")

        total_distance = 0
        current_pos = self.position.copy()

        for i, waypoint in enumerate(waypoints):
            x, y = waypoint
            distance = abs(x - current_pos[0]) + abs(y - current_pos[1])
            total_distance += distance
            current_pos = [x, y]
            print(f"  {i+1}ë‹¨ê³„: {waypoint} (ê±°ë¦¬: {distance})")

        print(f"ğŸ“ ì´ ê±°ë¦¬: {total_distance}")

        # ë°°í„°ë¦¬ë¡œ ê°ˆ ìˆ˜ ìˆëŠ”ì§€ í™•ì¸
        battery_needed = total_distance * 2
        if self.battery >= battery_needed:
            print("âœ… ë°°í„°ë¦¬ ì¶©ë¶„ - ê²½ë¡œ ì‹¤í–‰ ê°€ëŠ¥")
            return True
        else:
            print(f"âŒ ë°°í„°ë¦¬ ë¶€ì¡± - í•„ìš”: {battery_needed}%, í˜„ì¬: {self.battery}%")
            return False


class TrafficManager:
    """êµí†µ ê´€ë¦¬ ì‹œìŠ¤í…œ í´ë˜ìŠ¤"""

    def __init__(self):
        """êµí†µ ê´€ë¦¬ ì‹œìŠ¤í…œ ì´ˆê¸°í™”"""
        self.intersections = {
            'êµì°¨ë¡œA': {'signal': 'green', 'time_left': 30},
            'êµì°¨ë¡œB': {'signal': 'red', 'time_left': 45},
            'êµì°¨ë¡œC': {'signal': 'yellow', 'time_left': 5}
        }
        self.cars = []
        print("ğŸš¦ êµí†µ ê´€ë¦¬ ì‹œìŠ¤í…œ ì‹œì‘!")

    def add_car(self, car):
        """ì°¨ëŸ‰ ë“±ë¡"""
        self.cars.append(car)
        print(f"ğŸ“ {car.car_id} ì°¨ëŸ‰ ë“±ë¡ ì™„ë£Œ")

    def check_traffic_lights(self):
        """ì‹ í˜¸ë“± ìƒíƒœ í™•ì¸"""
        print("\nğŸš¦ í˜„ì¬ ì‹ í˜¸ë“± ìƒíƒœ:")

        for intersection, info in self.intersections.items():
            signal = info['signal']
            time_left = info['time_left']

            if signal == 'green':
                status = "ğŸŸ¢ í†µê³¼ ê°€ëŠ¥"
            elif signal == 'yellow':
                status = "ğŸŸ¡ ì£¼ì˜ í•„ìš”"
            else:
                status = "ğŸ”´ ì •ì§€ í•„ìš”"

            print(f"  {intersection}: {status} ({time_left}ì´ˆ)")

    def find_safe_route(self, start, end):
        """ì•ˆì „í•œ ê²½ë¡œ ì°¾ê¸°"""
        print(f"\nğŸ›£ï¸ {start} â†’ {end} ì•ˆì „ ê²½ë¡œ ê²€ìƒ‰")

        # ì‹ í˜¸ë“± ìƒíƒœ í™•ì¸í•´ì„œ ê²½ë¡œ ì¶”ì²œ
        green_lights = []
        red_lights = []

        for intersection, info in self.intersections.items():
            if info['signal'] == 'green':
                green_lights.append(intersection)
            elif info['signal'] == 'red':
                red_lights.append(intersection)

        if green_lights:
            recommended = green_lights[0]
            print(f"âœ… ì¶”ì²œ ê²½ë¡œ: {recommended} ê²½ìœ  (ì‹ í˜¸ ì–‘í˜¸)")
        else:
            print("âš ï¸ ëª¨ë“  ì‹ í˜¸ë“± ì£¼ì˜ - ì„œí–‰ ìš´ì „")

    def monitor_fleet(self):
        """ì°¨ëŸ‰ í•¨ëŒ€ ëª¨ë‹ˆí„°ë§"""
        print("\nğŸ“Š ì°¨ëŸ‰ í•¨ëŒ€ í˜„í™©:")

        if not self.cars:
            print("ë“±ë¡ëœ ì°¨ëŸ‰ì´ ì—†ìŠµë‹ˆë‹¤.")
            return

        total_cars = len(self.cars)
        low_battery_cars = []
        safe_cars = []

        for car in self.cars:
            battery_status = car.check_battery()
            sensor_ok = car.scan_sensors()

            print(f"\n{car.car_id}:")
            print(f"  ìœ„ì¹˜: {car.position}")
            print(f"  ë°°í„°ë¦¬: {car.battery}% ({battery_status})")
            print(f"  ì„¼ì„œ ìƒíƒœ: {'ì •ìƒ' if sensor_ok else 'ì ê²€í•„ìš”'}")

            if battery_status == "ë¶€ì¡±" or not sensor_ok:
                low_battery_cars.append(car.car_id)
            else:
                safe_cars.append(car.car_id)

        print(f"\nğŸ“ˆ í•¨ëŒ€ ìš”ì•½:")
        print(f"ì´ ì°¨ëŸ‰: {total_cars}ëŒ€")
        print(f"ì •ìƒ ìš´í–‰: {len(safe_cars)}ëŒ€")
        print(f"ì ê²€ í•„ìš”: {len(low_battery_cars)}ëŒ€")

        if low_battery_cars:
            print(f"âš ï¸ ì ê²€ ëŒ€ìƒ: {low_battery_cars}")


def simulate_emergency_scenario():
    """ì‘ê¸‰ ìƒí™© ì‹œë®¬ë ˆì´ì…˜ í•¨ìˆ˜"""
    print("\nğŸš¨ ì‘ê¸‰ ìƒí™© ë°œìƒ!")

    # ì‘ê¸‰ì°¨ëŸ‰ ìƒì„±
    ambulance = AutonomousCar("ì‘ê¸‰ì°¨_001", battery=80)

    # ì‘ê¸‰ ê²½ë¡œ ì„¤ì •
    emergency_route = [(5, 0), (5, 5), (10, 5), (10, 10)]

    print("ğŸ¥ ë³‘ì›ìœ¼ë¡œ ê¸´ê¸‰ ì´ì†¡ ì¤‘...")

    # ê²½ë¡œ ë”°ë¼ ì´ë™
    for i, waypoint in enumerate(emergency_route):
        x, y = waypoint
        print(f"\në‹¨ê³„ {i+1}: {waypoint}ë¡œ ì´ë™")

        # ì´ë™ ê°€ëŠ¥ ì—¬ë¶€ í™•ì¸
        if ambulance.move_to(x, y):
            print(f"âœ… ë‹¨ê³„ {i+1} ì™„ë£Œ")
        else:
            print(f"âŒ ë‹¨ê³„ {i+1} ì‹¤íŒ¨ - ì‘ê¸‰ ì¶©ì „ í•„ìš”")
            break

    print("ğŸ¥ ë³‘ì› ë„ì°©!")


def run_autonomous_driving_simulation():
    """ììœ¨ì£¼í–‰ ì‹œë®¬ë ˆì´ì…˜ ë©”ì¸ í•¨ìˆ˜"""
    print("=" * 50)
    print("ğŸš— ììœ¨ì£¼í–‰ì°¨ ì‹œë®¬ë ˆì´ì…˜ ì‹œì‘!")
    print("=" * 50)

    # êµí†µ ê´€ë¦¬ ì‹œìŠ¤í…œ ìƒì„±
    traffic_manager = TrafficManager()

    # ììœ¨ì£¼í–‰ì°¨ë“¤ ìƒì„±
    car1 = AutonomousCar("ììœ¨ì°¨_001", battery=90)
    car2 = AutonomousCar("ììœ¨ì°¨_002", battery=30)
    car3 = AutonomousCar("ììœ¨ì°¨_003", battery=75)

    # ì°¨ëŸ‰ë“¤ì„ êµí†µ ê´€ë¦¬ ì‹œìŠ¤í…œì— ë“±ë¡
    traffic_manager.add_car(car1)
    traffic_manager.add_car(car2)
    traffic_manager.add_car(car3)

    # ì‹ í˜¸ë“± í™•ì¸
    traffic_manager.check_traffic_lights()

    # ì•ˆì „ ê²½ë¡œ ê²€ìƒ‰
    traffic_manager.find_safe_route("ì¶œë°œì§€", "ëª©ì ì§€")

    # ì°¨ëŸ‰ë³„ ê²½ë¡œ ê³„íš ë° ì´ë™
    print("\n" + "=" * 30)
    print("ğŸ›£ï¸ ê°œë³„ ì°¨ëŸ‰ ìš´í–‰ ì‹œì‘")
    print("=" * 30)

    # ì°¨ëŸ‰ 1: ì •ìƒ ìš´í–‰
    route1 = [(3, 3), (6, 6), (10, 10)]
    if car1.plan_route(route1):
        for waypoint in route1:
            car1.move_to(waypoint[0], waypoint[1])

    # ì°¨ëŸ‰ 2: ë°°í„°ë¦¬ ë¶€ì¡±ìœ¼ë¡œ ì œí•œì  ì´ë™
    route2 = [(2, 2), (4, 4)]
    if car2.plan_route(route2):
        for waypoint in route2:
            if not car2.move_to(waypoint[0], waypoint[1]):
                break

    # ì°¨ëŸ‰ 3: ì„¼ì„œ ê³ ì¥ ì‹œë®¬ë ˆì´ì…˜
    car3.sensors['camera'] = False  # ì¹´ë©”ë¼ ê³ ì¥
    car3.sensors['lidar'] = False   # ë¼ì´ë‹¤ ê³ ì¥

    route3 = [(1, 1), (2, 2)]
    if car3.scan_sensors():  # ì„¼ì„œ í™•ì¸
        car3.plan_route(route3)
    else:
        print(f"âš ï¸ {car3.car_id}: ì„¼ì„œ ì´ìƒìœ¼ë¡œ ìš´í–‰ ì¤‘ë‹¨")

    # ì „ì²´ í•¨ëŒ€ ëª¨ë‹ˆí„°ë§
    traffic_manager.monitor_fleet()

    # ì‘ê¸‰ ìƒí™© ì‹œë®¬ë ˆì´ì…˜
    simulate_emergency_scenario()

    print("\n" + "=" * 50)
    print("ğŸ ì‹œë®¬ë ˆì´ì…˜ ì™„ë£Œ!")
    print("=" * 50)


# ì‹œë®¬ë ˆì´ì…˜ ì‹¤í–‰
if __name__ == "__main__":
    run_autonomous_driving_simulation()
```
