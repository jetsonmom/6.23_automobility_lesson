```python
# íŒŒì´ì¬ ì´ˆë³´ìë¥¼ ìœ„í•œ ììœ¨ì£¼í–‰ ê¸°ì´ˆ ì½”ë“œ
# ë³µì¡í•œ ìˆ˜í•™ ì—†ì´, ê°„ë‹¨í•œ ê°œë…ìœ¼ë¡œ ììœ¨ì£¼í–‰ì˜ í•µì‹¬ì„ ì´í•´í•´ë³´ì„¸ìš”!

print("ğŸš— íŒŒì´ì¬ìœ¼ë¡œ ë§Œë“œëŠ” ê°„ë‹¨í•œ ììœ¨ì£¼í–‰ì°¨!")
print("=" * 40)

# 1. ê°„ë‹¨í•œ ì°¨ëŸ‰ í´ë˜ìŠ¤
class SimpleCar:
    def __init__(self, name):
        self.name = name
        self.x = 0      # ì°¨ëŸ‰ì˜ x ìœ„ì¹˜
        self.y = 0      # ì°¨ëŸ‰ì˜ y ìœ„ì¹˜
        self.fuel = 100 # ì—°ë£Œ

    def move_right(self):
        """ì˜¤ë¥¸ìª½ìœ¼ë¡œ ì´ë™"""
        self.x += 1
        self.fuel -= 1
        print(f"{self.name}ì´(ê°€) ì˜¤ë¥¸ìª½ìœ¼ë¡œ ì´ë™! í˜„ì¬ ìœ„ì¹˜: ({self.x}, {self.y})")

    def move_up(self):
        """ìœ„ë¡œ ì´ë™"""
        self.y += 1
        self.fuel -= 1
        print(f"{self.name}ì´(ê°€) ìœ„ë¡œ ì´ë™! í˜„ì¬ ìœ„ì¹˜: ({self.x}, {self.y})")

    def show_status(self):
        """ì°¨ëŸ‰ ìƒíƒœ í‘œì‹œ"""
        print(f"ğŸš— {self.name} ìƒíƒœ:")
        print(f"   ìœ„ì¹˜: ({self.x}, {self.y})")
        print(f"   ì—°ë£Œ: {self.fuel}%")

# 2. ì¥ì• ë¬¼ ê°ì§€ í•¨ìˆ˜ (ë§¤ìš° ê°„ë‹¨!)
def detect_obstacle(car_x, car_y, obstacles):
    """ì°¨ëŸ‰ ì•ì— ì¥ì• ë¬¼ì´ ìˆëŠ”ì§€ í™•ì¸"""
    # ì¥ì• ë¬¼ ë¦¬ìŠ¤íŠ¸ì—ì„œ ì°¨ëŸ‰ ë°”ë¡œ ì• ìœ„ì¹˜ë¥¼ í™•ì¸
    next_position = (car_x + 1, car_y)  # ì˜¤ë¥¸ìª½ í•œ ì¹¸

    if next_position in obstacles:
        print("âš ï¸  ì•ì— ì¥ì• ë¬¼ ë°œê²¬!")
        return True
    else:
        print("âœ… ì•ì´ ì•ˆì „í•©ë‹ˆë‹¤.")
        return False

# 3. ê°„ë‹¨í•œ ìë™ ìš´ì „ í•¨ìˆ˜
def simple_autopilot(car, target_x, target_y, obstacles):
    """ëª©ì ì§€ê¹Œì§€ ìë™ìœ¼ë¡œ ìš´ì „í•˜ëŠ” í•¨ìˆ˜"""
    print(f"\nğŸ¯ ëª©ì ì§€: ({target_x}, {target_y})")
    print("ìë™ ìš´ì „ ì‹œì‘!")

    step = 1
    while car.x < target_x or car.y < target_y:
        print(f"\n--- {step}ë‹¨ê³„ ---")

        # ì—°ë£Œ ì²´í¬
        if car.fuel <= 0:
            print("â›½ ì—°ë£Œê°€ ë¶€ì¡±í•©ë‹ˆë‹¤!")
            break

        # xì¶•ìœ¼ë¡œ ë¨¼ì € ì´ë™
        if car.x < target_x:
            # ì¥ì• ë¬¼ ì²´í¬
            if not detect_obstacle(car.x, car.y, obstacles):
                car.move_right()
            else:
                # ì¥ì• ë¬¼ì´ ìˆìœ¼ë©´ ìœ„ë¡œ ìš°íšŒ
                print("ğŸ”„ ì¥ì• ë¬¼ ìš°íšŒ: ìœ„ë¡œ ì´ë™")
                car.move_up()

        # yì¶•ìœ¼ë¡œ ì´ë™
        elif car.y < target_y:
            car.move_up()

        step += 1

        # 10ë‹¨ê³„ë§ˆë‹¤ ìƒíƒœ í™•ì¸
        if step % 3 == 0:
            car.show_status()

    print(f"\nğŸ‰ ëª©ì ì§€ ë„ì°©! ì´ {step-1}ë‹¨ê³„ ì†Œìš”")

# 4. ê°„ë‹¨í•œ ì‹ í˜¸ë“± ì‹œë®¬ë ˆì´ì…˜
def traffic_light_system():
    """ì‹ í˜¸ë“± ì‹œìŠ¤í…œ ì‹œë®¬ë ˆì´ì…˜"""
    lights = ["ğŸ”´ ë¹¨ê°„ë¶ˆ", "ğŸŸ¡ ë…¸ë€ë¶ˆ", "ğŸŸ¢ ì´ˆë¡ë¶ˆ"]

    print("\nğŸš¦ ì‹ í˜¸ë“± ì‹œìŠ¤í…œ")
    for i in range(6):  # 6ë²ˆ ë³€ê²½
        current_light = lights[i % 3]
        print(f"í˜„ì¬ ì‹ í˜¸: {current_light}")

        if "ì´ˆë¡ë¶ˆ" in current_light:
            print("   â†’ ì¶œë°œ ê°€ëŠ¥!")
        elif "ë…¸ë€ë¶ˆ" in current_light:
            print("   â†’ ì£¼ì˜í•˜ì„¸ìš”!")
        else:
            print("   â†’ ì •ì§€!")
        print()

# 5. ì†ë„ ì œì–´ í•¨ìˆ˜
def speed_control(current_speed, speed_limit, obstacle_detected):
    """ì†ë„ë¥¼ ìë™ìœ¼ë¡œ ì¡°ì ˆí•˜ëŠ” í•¨ìˆ˜"""
    print(f"í˜„ì¬ ì†ë„: {current_speed}km/h")
    print(f"ì œí•œ ì†ë„: {speed_limit}km/h")

    if obstacle_detected:
        new_speed = 0
        print("ğŸš¨ ê¸´ê¸‰ ì •ì§€!")
    elif current_speed > speed_limit:
        new_speed = speed_limit
        print("â¬‡ï¸ ì†ë„ ê°ì†Œ")
    else:
        new_speed = min(current_speed + 10, speed_limit)
        print("â¬†ï¸ ì†ë„ ì¦ê°€")

    print(f"ì¡°ì •ëœ ì†ë„: {new_speed}km/h\n")
    return new_speed

# ì‹¤ì œ ì‹œë®¬ë ˆì´ì…˜ ì‹¤í–‰
if __name__ == "__main__":
    # ì°¨ëŸ‰ ìƒì„±
    my_car = SimpleCar("í…ŒìŠ¬ë¼")
    my_car.show_status()

    # ì¥ì• ë¬¼ ìœ„ì¹˜ ì„¤ì • (x, y ì¢Œí‘œ)
    obstacles = [(2, 0), (3, 0), (5, 1)]
    print(f"\nğŸš§ ì¥ì• ë¬¼ ìœ„ì¹˜: {obstacles}")

    # ìë™ ìš´ì „ ì‹œì‘
    simple_autopilot(my_car, target_x=6, target_y=3, obstacles=obstacles)

    print("\n" + "="*40)

    # ì‹ í˜¸ë“± ì‹œë®¬ë ˆì´ì…˜
    traffic_light_system()

    # ì†ë„ ì œì–´ ì‹œë®¬ë ˆì´ì…˜
    print("ğŸï¸ ì†ë„ ì œì–´ ì‹œë®¬ë ˆì´ì…˜")
    print("-" * 20)

    speed = 30
    for scenario in [
        (50, False, "ê³ ì†ë„ë¡œ ì§„ì…"),
        (80, False, "ì œí•œì†ë„ 80km/h êµ¬ê°„"),
        (80, True, "ì•ì°¨ ë°œê²¬!"),
        (60, False, "ì¥ì• ë¬¼ ì§€ë‚¨")
    ]:
        speed_limit, obstacle, situation = scenario
        print(f"ìƒí™©: {situation}")
        speed = speed_control(speed, speed_limit, obstacle)

    print("ğŸ“ íŒŒì´ì¬ìœ¼ë¡œ ììœ¨ì£¼í–‰ì˜ ê¸°ë³¸ ì›ë¦¬ë¥¼ ë°°ì› ìŠµë‹ˆë‹¤!")
    print("   - ê°ì²´ì§€í–¥ í”„ë¡œê·¸ë˜ë° (í´ë˜ìŠ¤)")
    print("   - ì¡°ê±´ë¬¸ê³¼ ë°˜ë³µë¬¸")
    print("   - í•¨ìˆ˜ ì •ì˜ì™€ í™œìš©")
    print("   - ë¦¬ìŠ¤íŠ¸ì™€ íŠœí”Œ ì‚¬ìš©")
    print("   - ê°„ë‹¨í•œ ì•Œê³ ë¦¬ì¦˜ êµ¬í˜„")

    print("\nğŸ’¡ ì´ê²ƒì´ ë°”ë¡œ íŒŒì´ì¬ì˜ í˜ì…ë‹ˆë‹¤!")
    print("   ë³µì¡í•œ ììœ¨ì£¼í–‰ë„ ì´ë ‡ê²Œ ê°„ë‹¨í•˜ê²Œ ì‹œì‘í•  ìˆ˜ ìˆì–´ìš”!")
```
