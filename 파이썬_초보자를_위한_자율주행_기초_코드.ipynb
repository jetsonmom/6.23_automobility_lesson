```python
# 파이썬 초보자를 위한 자율주행 기초 코드
# 복잡한 수학 없이, 간단한 개념으로 자율주행의 핵심을 이해해보세요!

print("🚗 파이썬으로 만드는 간단한 자율주행차!")
print("=" * 40)

# 1. 간단한 차량 클래스
class SimpleCar:
    def __init__(self, name):
        self.name = name
        self.x = 0      # 차량의 x 위치
        self.y = 0      # 차량의 y 위치
        self.fuel = 100 # 연료

    def move_right(self):
        """오른쪽으로 이동"""
        self.x += 1
        self.fuel -= 1
        print(f"{self.name}이(가) 오른쪽으로 이동! 현재 위치: ({self.x}, {self.y})")

    def move_up(self):
        """위로 이동"""
        self.y += 1
        self.fuel -= 1
        print(f"{self.name}이(가) 위로 이동! 현재 위치: ({self.x}, {self.y})")

    def show_status(self):
        """차량 상태 표시"""
        print(f"🚗 {self.name} 상태:")
        print(f"   위치: ({self.x}, {self.y})")
        print(f"   연료: {self.fuel}%")

# 2. 장애물 감지 함수 (매우 간단!)
def detect_obstacle(car_x, car_y, obstacles):
    """차량 앞에 장애물이 있는지 확인"""
    # 장애물 리스트에서 차량 바로 앞 위치를 확인
    next_position = (car_x + 1, car_y)  # 오른쪽 한 칸

    if next_position in obstacles:
        print("⚠️  앞에 장애물 발견!")
        return True
    else:
        print("✅ 앞이 안전합니다.")
        return False

# 3. 간단한 자동 운전 함수
def simple_autopilot(car, target_x, target_y, obstacles):
    """목적지까지 자동으로 운전하는 함수"""
    print(f"\n🎯 목적지: ({target_x}, {target_y})")
    print("자동 운전 시작!")

    step = 1
    while car.x < target_x or car.y < target_y:
        print(f"\n--- {step}단계 ---")

        # 연료 체크
        if car.fuel <= 0:
            print("⛽ 연료가 부족합니다!")
            break

        # x축으로 먼저 이동
        if car.x < target_x:
            # 장애물 체크
            if not detect_obstacle(car.x, car.y, obstacles):
                car.move_right()
            else:
                # 장애물이 있으면 위로 우회
                print("🔄 장애물 우회: 위로 이동")
                car.move_up()

        # y축으로 이동
        elif car.y < target_y:
            car.move_up()

        step += 1

        # 10단계마다 상태 확인
        if step % 3 == 0:
            car.show_status()

    print(f"\n🎉 목적지 도착! 총 {step-1}단계 소요")

# 4. 간단한 신호등 시뮬레이션
def traffic_light_system():
    """신호등 시스템 시뮬레이션"""
    lights = ["🔴 빨간불", "🟡 노란불", "🟢 초록불"]

    print("\n🚦 신호등 시스템")
    for i in range(6):  # 6번 변경
        current_light = lights[i % 3]
        print(f"현재 신호: {current_light}")

        if "초록불" in current_light:
            print("   → 출발 가능!")
        elif "노란불" in current_light:
            print("   → 주의하세요!")
        else:
            print("   → 정지!")
        print()

# 5. 속도 제어 함수
def speed_control(current_speed, speed_limit, obstacle_detected):
    """속도를 자동으로 조절하는 함수"""
    print(f"현재 속도: {current_speed}km/h")
    print(f"제한 속도: {speed_limit}km/h")

    if obstacle_detected:
        new_speed = 0
        print("🚨 긴급 정지!")
    elif current_speed > speed_limit:
        new_speed = speed_limit
        print("⬇️ 속도 감소")
    else:
        new_speed = min(current_speed + 10, speed_limit)
        print("⬆️ 속도 증가")

    print(f"조정된 속도: {new_speed}km/h\n")
    return new_speed

# 실제 시뮬레이션 실행
if __name__ == "__main__":
    # 차량 생성
    my_car = SimpleCar("테슬라")
    my_car.show_status()

    # 장애물 위치 설정 (x, y 좌표)
    obstacles = [(2, 0), (3, 0), (5, 1)]
    print(f"\n🚧 장애물 위치: {obstacles}")

    # 자동 운전 시작
    simple_autopilot(my_car, target_x=6, target_y=3, obstacles=obstacles)

    print("\n" + "="*40)

    # 신호등 시뮬레이션
    traffic_light_system()

    # 속도 제어 시뮬레이션
    print("🏎️ 속도 제어 시뮬레이션")
    print("-" * 20)

    speed = 30
    for scenario in [
        (50, False, "고속도로 진입"),
        (80, False, "제한속도 80km/h 구간"),
        (80, True, "앞차 발견!"),
        (60, False, "장애물 지남")
    ]:
        speed_limit, obstacle, situation = scenario
        print(f"상황: {situation}")
        speed = speed_control(speed, speed_limit, obstacle)

    print("🎓 파이썬으로 자율주행의 기본 원리를 배웠습니다!")
    print("   - 객체지향 프로그래밍 (클래스)")
    print("   - 조건문과 반복문")
    print("   - 함수 정의와 활용")
    print("   - 리스트와 튜플 사용")
    print("   - 간단한 알고리즘 구현")

    print("\n💡 이것이 바로 파이썬의 힘입니다!")
    print("   복잡한 자율주행도 이렇게 간단하게 시작할 수 있어요!")
```
